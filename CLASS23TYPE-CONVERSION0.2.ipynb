```python
#BODMAS Rule (Order of Operations): BODMAS tells us the correct order to solve mathematical expressions.

#BODMAS Full Form
Letter	Meaning
B	    Brackets
O	    Orders (Powers, Roots)
D	    Division
M	    Multiplication
A	    Addition
S	    Subtraction



#Correct Order

1️⃣ Brackets
2️⃣ Orders (powers like square, cube)
3️⃣ Division & Multiplication (left to right)
4️⃣ Addition & Subtraction (left to right)
```

```python
Example 1
8+2∗3


Step-by-step:
Multiply first → 2 × 3 = 6
Then add → 8 + 6 = 14
✔ Final Answer = 14
```

```python
Example 2
(8+2)∗3


Step-by-step:
Bracket first → 8 + 2 = 10
Then multiply → 10 × 3 = 30
✔ Final Answer = 30
```

```python
Example 3 (With Power)
5+23∗25+23∗2


Step-by-step:
Order (power) → 2³ = 8
Multiply → 8 × 2 = 16
Add → 5 + 16 = 21

✔ Final Answer = 21
```

```python
Important Rule

👉 Multiplication and Division are equal priority
👉 Addition and Subtraction are equal priority
👉 Solve from LEFT TO RIGHT

Example:

20/5∗2
20 ÷ 5 = 4
4 × 2 = 8

✔ Answer = 8
```

```python
#BODMAS in Python:--Python follows BODMAS automatically:

print(8 + 2 * 3)     # 14
print((8 + 2) * 3)   # 30
print(5 + 2**3 * 2)  # 21
```

```python
#1. Simple Arithmetic

#10+5∗2

print(10 + 5 * 2)



# Explanation:
# Multiply first → 5 × 2 = 10
# Add → 10 + 10 = 20

# ✔ Output = 20
```

```python
#2. Using Brackets

#(10+5)∗2
print((10 + 5) * 2)

# Bracket first → 15
# Multiply → 15 × 2 = 30

# ✔ Output = 30
```

```python
#3. With Power (Order)

#2+3∗∗2∗4

print(2 + 3**2 * 4)

# Step-by-step:
# Power → 3² = 9
# Multiply → 9 × 4 = 36
# Add → 2 + 36 = 38

# ✔ Output = 38
```

```python
#4. Division & Multiplication (Left to Right)

#100/5∗2
print(100 / 5 * 2)

# 100 ÷ 5 = 20
# 20 × 2 = 40

# ✔ Output = 40
```

```python
#5. Real-Life Programming Example (Student Percentage)

# (450+10)/500∗100


marks = 450
bonus = 10
total = 500

percentage = (marks + bonus) / total * 100
print(percentage)

# ✔ Output = 92.0
```

```python
#6. Complex Expression in Code

# 5+6∗2−8/4

print(5 + 6 * 2 - 8 / 4)

# Step-by-step:
# Multiply → 6 × 2 = 12
# Divide → 8 ÷ 4 = 2
# Add → 5 + 12 = 17
# Subtract → 17 − 2 = 15

# ✔ Output = 15
```

```python
# Important Programming Note

# In Python:
# ** → Power
# / → Division
# // → Floor Division
# % → Modulus

# Example:

print(10 // 3)   # 3
print(10 % 3)    # 1
```

```python
#1. Student Percentage Calculation

# Formula:(marks/total)∗100


marks = 450
total = 500

percentage = (marks / total) * 100
print("Percentage:", percentage)

# ✔ Bracket first
# ✔ Division
# ✔ Multiplication
```

```python
#2. Simple Interest Calculator

#Formula:(P∗R∗T)/100


P = 10000   # Principal
R = 5       # Rate
T = 2       # Time

SI = (P * R * T) / 100
print("Simple Interest:", SI)


# ✔ Multiplication first
# ✔ Then division
```

```python
#3.Area of Circle

# Formula:3.14∗r∗∗2


r = 7
area = 3.14 * r**2
print("Area:", area)

# ✔ Power first
# ✔ Then multiplication
```

```python
#4. EMI Calculation (Banking App)

# Simplified Formula:(P∗R∗(1+R)∗∗N)/((1+R)∗∗N−1)

P = 100000
R = 0.01
N = 12

emi = (P * R * (1+R)**N) / ((1+R)**N - 1)
print("EMI:", emi)



# ✔ Power
# ✔ Multiplication
# ✔ Division
# ✔ Subtraction

# All follow BODMAS automatically.
```

```python
#5. Game Score Calculation


# bases​core+level∗bonus−penalty/2


base_score = 100
level = 5
bonus = 20
penalty = 10

final_score = base_score + level * bonus - penalty / 2
print("Final Score:", final_score)



# ✔ Multiply first
# ✔ Divide
# ✔ Add
# ✔ Subtract
```

```python
#Mini Calculator Project (Python)
Let’s build a real working calculator step-by-step 🚀

#This project will use:
✔ Type Conversion
✔ BODMAS
✔ Conditional statements
✔ Functions
✔ Loop
```

```python
#Version 1 – Simple Calculator

def calculator():
    print("Mini Calculator")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    choice = input("Choose (1-4): ")

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == '1':
        print("Result:", num1 + num2)

    elif choice == '2':
        print("Result:", num1 - num2)

    elif choice == '3':
        print("Result:", num1 * num2)

    elif choice == '4':
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Cannot divide by zero")

    else:
        print("Invalid choice")

calculator()
```
