```python
# -- BBB-- ALL '.' format()

The .format() method is used for string formatting.
It replaces {} placeholders inside a string with values.

Syntax:

"string {}".format(value)
```

```python
# Single value
name = "Vishal"
print("Hello {}".format(name))
```

```python
# Multiple values
name = "Vishal"
age = 21
print("Name: {} Age: {}".format(name, age))
```

```python
# Positional Formatting

# You can specify positions inside {}.

print("{0} is {1} years old".format("Vishal", 21))
```

```python
print("{1} is {0} years old".format(21, "Vishal"))
```

```python
# Keyword Formatting
print("Name: {n} Age: {a}".format(n="Vishal", a=21))
```

```python
# Format Specifiers with .format()
# Integer

print("Number: {:d}".format(10))

# Float
print("Price: {:f}".format(25.5))
```

```python
# Float with precision
print("Pi: {:.2f}".format(3.14159))
```

```python
# Width and Alignment
# Width
print("{:5d}".format(7))
```

```python
# Left alignment
print("{:<5}".format("Hi"))
```

```python
# Right alignment
print("{:>5}".format("Hi"))
```

```python
# Center alignment
print("{:^5}".format("Hi"))
```

```python
# Zero Padding
print("{:05d}".format(42))
```

```python
# Common Type Specifiers
# Specifier	Meaning	Example
# d	Integer (decimal)	{:d}
# f	Float	{:f}
# s	String	{:s}
# b	Binary	{:b}
# o	Octal	{:o}
# x	Hex (lowercase)	{:x}
# X	Hex (uppercase)	{:X}
# %	Percentage	{:%}


# Syntax for Specifying Type in .format()
General Syntax
"{:format_specifier}".format(value)


# Full Syntax Structure
"{: [fill] [align] [width] [.precision] [type] }".format(value)


Parts:
1.fill → character used to fill empty space
2.align → alignment of text
3.width → total width of output
4.precision → number of digits after decimal
5.type → data type specifier


# Alignment Symbols
Symbol	    Meaning
1.<	        Left align
2.>	        Right align
3.^	        Center align
```

```python
# -- CCC -- f' format()

Definition :- An f-string is a modern and easy way to format strings in Python.
It allows you to insert variables or expressions directly inside a string using curly braces {}.

Basic Syntax
f"string {variable}"
```

```python
# Examples
# 1. Using a variable

name = "Vishal"
print(f"Hello {name}")
```

```python
# 2. Multiple variables

name = "Vishal"
age = 21
print(f"Name: {name}, Age: {age}")
```

```python
# 3. Expressions inside f-string

a = 5
b = 3
print(f"Sum: {a + b}")
```

```python
# 4. Float formatting

pi = 3.14159
print(f"Pi: {pi:.2f}")
```

```python
# 5. Alignment in f-strings

text = "Hi"
print(f"{text:<10}")   # left align
print(f"{text:>10}")   # right align
print(f"{text:^10}")   # center align
```

```python
# 6. Zero padding

num = 42
print(f"{num:05d}")
```

```python
Common Type Specifiers in f-strings
Specifier	Meaning	Example
d	Integer	{num:d}
f	Float	{pi:.2f}
s	String	{name:s}
b	Binary	{num:b}
o	Octal	{num:o}
x	Hex (lower)	{num:x}
%	Percentage	{val:.2%}


Important Points
Always start the string with f or F.
Variables are written inside {}.
You can use expressions, functions, or formatting inside {}.
```

```python
# -- EXAMPLE -- Here is a big, real-life example using f-strings that includes:
1.variables
2.calculations
3.alignment
4.width
5.precision
6.different number formats

# -- Big Example: Student Report Card
```

```python
name = "Vishal"
roll_no = 23
math = 85
science = 78
english = 92

# Calculations
total = math + science + english
average = total / 3

# Printing formatted report
print(f"{'Student Report':^40}")
print("-" * 40)

print(f"Name      : {name}")
print(f"Roll No   : {roll_no:03d}")

print("-" * 40)
print(f"{'Subject':<15}{'Marks':>10}")
print("-" * 40)

print(f"{'Math':<15}{math:>10d}")
print(f"{'Science':<15}{science:>10d}")
print(f"{'English':<15}{english:>10d}")

print("-" * 40)
print(f"{'Total':<15}{total:>10d}")
print(f"{'Average':<15}{average:>10.2f}")
print("-" * 40)
```

```python
# -- What this example shows

| Feature          | Example                            |
| ---------------- | ---------------------------------- |
| Center alignment | `{'Student Report':^40}`           |
| Left alignment   | `{'Math':<15}`                     |
| Right alignment  | `{math:>10d}`                      |
| Zero padding     | `{roll_no:03d}`                    |
| Float precision  | `{average:.2f}`                    |
| Expressions      | `total = math + science + english` |
```

```python
shop_name = "Vishal Super Store"

# Item details
item1 = "Rice"
price1 = 60
qty1 = 3

item2 = "Milk"
price2 = 45
qty2 = 2

item3 = "Bread"
price3 = 30
qty3 = 4

# Calculations
total1 = price1 * qty1
total2 = price2 * qty2
total3 = price3 * qty3

subtotal = total1 + total2 + total3
tax = subtotal * 0.05
grand_total = subtotal + tax

# Bill printing
print(f"{shop_name:^40}")
print("-" * 40)
print(f"{'Item':<15}{'Qty':>5}{'Price':>10}{'Total':>10}")
print("-" * 40)

print(f"{item1:<15}{qty1:>5}{price1:>10.2f}{total1:>10.2f}")
print(f"{item2:<15}{qty2:>5}{price2:>10.2f}{total2:>10.2f}")
print(f"{item3:<15}{qty3:>5}{price3:>10.2f}{total3:>10.2f}")

print("-" * 40)
print(f"{'Subtotal':<30}{subtotal:>10.2f}")
print(f"{'Tax (5%)':<30}{tax:>10.2f}")
print("-" * 40)
print(f"{'Grand Total':<30}{grand_total:>10.2f}")
print("-" * 40)
```

```python
# Here is a **detailed explanation** of the **ATM receipt example** step by step.

#EXAMPLE
# ATM Receipt Program


bank_name = "State Bank of India"
location = "Pune Branch"

name = "Vishal"
account_no = 1234567890
transaction_type = "Withdrawal"
amount = 2500.75

balance_before = 10000.50
balance_after = balance_before - amount

print(f"{bank_name:^40}")
print(f"{location:^40}")
print("-" * 40)

print(f"Name            : {name}")
print(f"Account No      : {account_no}")
print(f"Transaction     : {transaction_type}")

print("-" * 40)
print(f"{'Amount':<25}: {amount:>10.2f}")
print(f"{'Balance Before':<25}: {balance_before:>10.2f}")
print(f"{'Balance After':<25}: {balance_after:>10.2f}")
print("-" * 40)

print(f"{'Thank You! Visit Again':^40}")
print("-" * 40)





# --  Step-by-Step Explanation
### 1. Bank Details
bank_name = "State Bank of India"
location = "Pune Branch"

#These variables store the **bank name** and **branch location**.


### 2. Customer Details
name = "Vishal"
account_no = 1234567890
transaction_type = "Withdrawal"
amount = 2500.75

#These variables store:

* Customer name
* Account number
* Type of transaction
* Withdrawn amount


### 3. Balance Calculation
balance_before = 10000.50
balance_after = balance_before - amount

Here:
* `balance_before` = money in account before withdrawal
* `balance_after` = remaining balance after subtracting amount

Example:
10000.50 − 2500.75 = 7499.75


### 4. Printing Bank Header (Center Alignment)
print(f"{bank_name:^40}")
print(f"{location:^40}")

* `^` means **center alignment**
* `40` is the total width

#So the text appears **in the center** of the line.


### 5. Printing Separator Line
print("-" * 40)

* Prints `-` 40 times
* Used to create a border line


### 6. Printing Customer Information
print(f"Name            : {name}")
print(f"Account No      : {account_no}")
print(f"Transaction     : {transaction_type}")

#This prints simple text with values.


### 7. Printing Amount Section
print(f"{'Amount':<25}: {amount:>10.2f}")

Breakdown:
A.`'Amount':<25`
  * `<` means **left alignment**
  * `25` is width


B. `amount:>10.2f`
  * `>` means **right alignment**
  * `10` is width
  * `.2f` means **2 decimal places**

      
#So the number is printed like:
    2500.75


      
### 8. Balance Lines
print(f"{'Balance Before':<25}: {balance_before:>10.2f}")
print(f"{'Balance After':<25}: {balance_after:>10.2f}")

#Same formatting rules:
* Left-aligned labels
* Right-aligned numbers
* Two decimal places


### 9. Thank You Message
print(f"{'Thank You! Visit Again':^40}")


* Center aligned message
* Width = 40


## Key Formatting Rules Used

| Format     | Meaning                     |
| ---------- | --------------------------- |
| `^40`      | Center align in width 40    |
| `<25`      | Left align in width 25      |
| `>10`      | Right align in width 10     |
| `.2f`      | Float with 2 decimal places |
| `"-" * 40` | Repeat `-` 40 times         |


## Flow of the Program

1. Store bank and customer details.
2. Calculate balance.
3. Print header.
4. Print transaction details.
5. Print balances.
6. Print thank-you message.
```
