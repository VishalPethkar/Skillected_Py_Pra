```python
#-- TYPE CONVERSION FUNCTIONS IN PYTHON (FULL DETAIL)

-- Type conversion means converting one data type into another.

#--There are 2 types:


#I. Implicit Type Conversion (Automatic)
Python automatically converts smaller data type to bigger type.
```

```python
#Example:
a = 10      # int
b = 5.5     # float
c = a + b
print(c)
print(type(c))
```

```python
#II. Explicit Type Conversion (Manual)
We convert data type using functions.

#QUICK SUMMARY TABLE
Function	Converts To
int()	Integer
float()	Float
complex()	Complex
str()	String
bool()	Boolean
list()	List
set()	Set
tuple()	Tuple
dict()	Dictionary
eval()	Expression
repr()	String representation
bin()	Binary
oct()	Octal
hex()	Hex
ord()	ASCII number
chr()	Character
```

```python
#-- Examples--

#1. int():Converts to integer.
print(int(10.9))     # 10
print(int("25"))     # 25




#2. float():Converts to float.
print(float(10))      # 10.0
print(float("5.6"))   # 5.6




#3. complex():Converts to complex number.
print(complex(5))        # (5+0j)
print(complex(2,3))      # (2+3j)




#4. str():Converts to string.
print(str(100))    # "100"




#5. bool():Rules:
# 0 → False
# Empty → False
# Others → True

print(bool(0))        # False
print(bool(10))       # True
print(bool(""))       # False




#6. list()
print(list("abc"))     # ['a','b','c']
print(list((1,2,3)))   # [1,2,3]




#7. set():Removes duplicates.
print(set([1,2,2,3]))   # {1,2,3}




#8. tuple()
print(tuple([1,2,3]))   # (1,2,3)




#9. dict():Needs key-value pairs.
print(dict([(1,'a'), (2,'b')]))




#10. eval():converts string expression to real expression.
print(eval("10+5"))   # 15

#⚠ Dangerous if used with user input.





#11. repr():Returns string representation of object.
x = "Hello"
print(repr(x))




#12. bin():Binary conversion.
print(bin(10))   # 0b1010




#13. oct():Octal conversion.
print(oct(10))   # 0o12




#14. hex():Hexadecimal conversion.
print(hex(10))   # 0xa




#15. ord():Gives ASCII value.
print(ord('A'))   # 65




#16. chr() (Not char):Returns character from ASCII value.
print(chr(65))   # A
```

```python
#-- Example-- Real-Life Mini Project

Student Result & Data Conversion System
This project simulates a school result processing system and uses almost all type conversion functions.
```

```python
#Project Goal

✔ Take student data (mostly strings)
✔ Convert to proper data types
✔ Calculate result
✔ Show binary, octal, hex values
✔ Show ASCII values
✔ Use list, set, tuple, dict conversions
```

```python
#Step 1: Raw Input Data (All Strings)
student_data = {
    "roll": "101",
    "name": "Vish",
    "marks": "450",
    "total": "500",
    "bonus": "5",
    "subjects": "Math,Science,English,Math"
}
```

```python
#Step 2: Apply Conversions

# 1. int()
roll = int(student_data["roll"])
marks = int(student_data["marks"])
total = int(student_data["total"])
bonus = int(student_data["bonus"])

# 2. float()
percentage = float(marks + bonus) / total * 100

# 3. complex()
complex_marks = complex(marks)

# 4. str()
result_text = str(percentage) + "%"

# 5. bool()
is_pass = bool(percentage >= 35)

# 6. list()
subject_list = list(student_data["subjects"].split(","))

# 7. set()
unique_subjects = set(subject_list)

# 8. tuple()
subject_tuple = tuple(unique_subjects)

# 9. dict()
subject_dict = dict.fromkeys(unique_subjects, 0)

# 10. eval()
extra_marks = eval("10 + 5")

# 11. repr()
repr_name = repr(student_data["name"])

# 12. bin()
binary_roll = bin(roll)

# 13. oct()
oct_roll = oct(roll)

# 14. hex()
hex_roll = hex(roll)

# 15. ord()
ascii_first_letter = ord(student_data["name"][0])

# 16. chr()
character_from_ascii = chr(65)
```

```python
#-- Output--

print("Roll:", roll)
print("Binary Roll:", binary_roll)
print("Octal Roll:", oct_roll)
print("Hex Roll:", hex_roll)

print("Percentage:", percentage)
print("Result Text:", result_text)
print("Pass:", is_pass)

print("Subjects List:", subject_list)
print("Unique Subjects:", unique_subjects)
print("Tuple Subjects:", subject_tuple)
print("Subject Dictionary:", subject_dict)

print("Complex Marks:", complex_marks)
print("Extra Marks using eval:", extra_marks)

print("repr Name:", repr_name)
print("ASCII of First Letter:", ascii_first_letter)
print("Character from ASCII 65:", character_from_ascii)
```

```python
#--What This Project Teaches
Conversion	               Real Usage
int()	                   Convert marks from string
float()	                   Calculate percentage
complex()	               Store complex marks
str()	                   Format output
bool()	                   Check pass/fail
list()	                   Convert subjects to list
set()	                   Remove duplicates
tuple()                	   Make subjects fixed
dict()	                   Initialize subject marks
eval()	                   Evaluate math expression
repr()	                   Debug string
bin/oct/hex	               Number system conversion
ord/chr	                   ASCII operations
```
