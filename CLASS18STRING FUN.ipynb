````python
# String Functions in Python
# Strings in Python have many built-in functions (methods) that start with a dot (.) and are used to perform operations on strings.

Syntax:
string.method()
````

````python
# Example:

text = "hello"
print(text.upper())
````

````python
# 1. capitalize()
# Makes first letter uppercase, rest lowercase

text = "hello WORLD"
print(text.capitalize())
````

````python
# 2. upper()
# Converts string to uppercase

print("hello".upper())
````

````python
# 3. lower()
# Converts string to lowercase

print("HELLO".lower())
````

````python
# 4. title()
# Capitalizes first letter of each word

print("hello world".title())
````

````python
# 5. swapcase()
# Swaps uppercase to lowercase and vice versa

print("HeLLo".swapcase())
````

````python
# 6. isupper()
# Checks if all letters are uppercase

print("HELLO".isupper())
````

````python
# 7. islower()
# Checks if all letters are lowercase

print("hello".islower())
````

````python
# 8. len()
# Returns length of string

print(len("hello"))
````

````python
# 9. max(str)
# Returns highest ASCII character

print(max("apple"))
````

````python
# 10. min(str)
# Returns lowest ASCII character

print(min("apple"))
````

````python
# 11. count()
# Counts occurrences of substring

print("banana".count("a"))
````

````python
# 12. center()
# Centers string inside given width

print("Hi".center(10, "*"))
````

````python
# 13. startswith()
# Checks starting text

print("hello".startswith("he"))
````

````python
# 14. endswith()
# Checks ending text

print("hello".endswith("lo"))
````

````python
# 15. isdigit()
# Checks if all characters are digits

print("12345".isdigit())
````

````python
# 16. isdecimal()
# Checks if all characters are decimal digits

print("123".isdecimal())
````

````python
# 17. isalnum()
# Checks if letters + numbers only

print("abc123".isalnum())
````

````python
# 18. isascii()
# Checks if all characters are ASCII

print("hello".isascii())
````

````python
# 19. isalpha()
# Checks if all letters

print("hello".isalpha())
````

````python
# 20. index()
# Finds position of substring (error if not found)

print("banana".index("n"))
````

````python
# 21. find()
# Finds position of substring (-1 if not found)

print("banana".find("n"))
````

````python
# 22. rindex()
# Finds last position (error if not found)

print("banana".rindex("a"))
````

````python
# 23. rfind()
# Finds last position (-1 if not found)

print("banana".rfind("a"))
````

````python
# 24. join()
# Joins list into string

words = ["apple", "banana", "mango"]
print("-".join(words))
````

````python
# 25. rjust()
# Right-aligns string

print("Hi".rjust(10, "*"))
````

````python
# 26. ljust()
# Left-aligns string

print("Hi".ljust(10, "*"))
````

````python
# 27. strip()
# Removes spaces from both sides

print("  hello  ".strip())
````

````python
# 28. rstrip()
# Removes spaces from right

print("hello  ".rstrip())
````

````python
# 29. lstrip()
# Removes spaces from left

print("  hello".lstrip())
````

````python
# 30. split()
# Splits string into list

print("apple mango banana".split())
````

````python
# 31. partition()
# Splits into 3 parts at first occurrence

print("hello-world".partition("-"))
````

````python
# 32. rpartition()
# Splits into 3 parts at last occurrence

print("one-two-three".rpartition("-"))
````

````python
# Big Program: Text Analyzer Using String Functions

This program:
1.Takes a sentence
2.Applies many string functions
3.Shows different results
````

````python
# -- EXAMPLE
text = "   Hello Vishal123 welcome-to PYTHON programming!!!   "

print("Original text:")
print(text)

# Strip spaces
clean = text.strip()
print("\nAfter strip():", clean)

# Case conversions
print("\nUpper:", clean.upper())
print("Lower:", clean.lower())
print("Title:", clean.title())
print("Swapcase:", clean.swapcase())
print("Capitalize:", clean.capitalize())

# Checking functions
print("\nIs upper:", clean.isupper())
print("Is lower:", clean.islower())
print("Is alpha:", clean.isalpha())
print("Is digit:", clean.isdigit())
print("Is decimal:", clean.isdecimal())
print("Is alnum:", clean.isalnum())
print("Is ascii:", clean.isascii())

# Length, max, min
print("\nLength:", len(clean))
print("Max char:", max(clean))
print("Min char:", min(clean))

# Count
print("\nCount of 'o':", clean.count("o"))

# Find and index
print("\nFirst 'o' using find():", clean.find("o"))
print("Last 'o' using rfind():", clean.rfind("o"))

# Starts and ends
print("\nStarts with 'Hello':", clean.startswith("Hello"))
print("Ends with '!!!':", clean.endswith("!!!"))

# Center, ljust, rjust
print("\nCenter:")
print(clean.center(60, "*"))

print("\nLeft Justify:")
print(clean.ljust(60, "-"))

print("\nRight Justify:")
print(clean.rjust(60, "-"))

# Replace
replaced = clean.replace("PYTHON", "Java")
print("\nAfter replace():", replaced)

# Split
words = clean.split()
print("\nSplit into words:", words)

# Join
joined = "_".join(words)
print("Joined with _ :", joined)

# Partition
print("\nPartition at '-':", clean.partition("-"))
print("Rpartition at '-':", clean.rpartition("-"))

# Index and rindex (only if character exists)
if "o" in clean:
    print("\nIndex of 'o':", clean.index("o"))
    print("Last index of 'o':", clean.rindex("o"))
````

````python
## Python String Functions – One-Page Cheat Sheet

### Case Conversion

| Function       | Description                         | Example                 | Output          |
| -------------- | ----------------------------------- | ----------------------- | --------------- |
| `capitalize()` | First letter uppercase              | `"hello".capitalize()`  | `"Hello"`       |
| `upper()`      | All letters uppercase               | `"hello".upper()`       | `"HELLO"`       |
| `lower()`      | All letters lowercase               | `"HELLO".lower()`       | `"hello"`       |
| `title()`      | First letter of each word uppercase | `"hello world".title()` | `"Hello World"` |
| `swapcase()`   | Swap upper ↔ lower                  | `"HeLLo".swapcase()`    | `"hEllO"`       |

---

### Checking Functions (Return True/False)

| Function      | Meaning                | Example              | Output |
| ------------- | ---------------------- | -------------------- | ------ |
| `isupper()`   | All uppercase?         | `"HELLO".isupper()`  | `True` |
| `islower()`   | All lowercase?         | `"hello".islower()`  | `True` |
| `isdigit()`   | Only digits?           | `"123".isdigit()`    | `True` |
| `isdecimal()` | Decimal numbers only   | `"123".isdecimal()`  | `True` |
| `isalnum()`   | Letters + numbers only | `"abc123".isalnum()` | `True` |
| `isalpha()`   | Only letters           | `"abc".isalpha()`    | `True` |
| `isascii()`   | ASCII characters only  | `"abc".isascii()`    | `True` |

---

### Length, Max, Min

| Function   | Description       | Example        | Output |
| ---------- | ----------------- | -------------- | ------ |
| `len(str)` | Length of string  | `len("hello")` | `5`    |
| `max(str)` | Highest character | `max("abc")`   | `'c'`  |
| `min(str)` | Lowest character  | `min("abc")`   | `'a'`  |

---

### Searching Functions

| Function       | Description                      | Example                    | Output |
| -------------- | -------------------------------- | -------------------------- | ------ |
| `count()`      | Count substring                  | `"banana".count("a")`      | `3`    |
| `find()`       | First index or -1                | `"hello".find("l")`        | `2`    |
| `rfind()`      | Last index or -1                 | `"hello".rfind("l")`       | `3`    |
| `index()`      | First index (error if not found) | `"hello".index("l")`       | `2`    |
| `rindex()`     | Last index (error if not found)  | `"hello".rindex("l")`      | `3`    |
| `startswith()` | Starts with substring?           | `"hello".startswith("he")` | `True` |
| `endswith()`   | Ends with substring?             | `"hello".endswith("lo")`   | `True` |

---

### Alignment & Spacing

| Function        | Description  | Example               | Output     |
| --------------- | ------------ | --------------------- | ---------- |
| `center(width)` | Center align | `"hi".center(6, "-")` | `"--hi--"` |
| `ljust(width)`  | Left align   | `"hi".ljust(6, "-")`  | `"hi----"` |
| `rjust(width)`  | Right align  | `"hi".rjust(6, "-")`  | `"----hi"` |

---

### Trimming Functions

| Function   | Description              | Example          | Output |
| ---------- | ------------------------ | ---------------- | ------ |
| `strip()`  | Remove spaces both sides | `" hi ".strip()` | `"hi"` |
| `lstrip()` | Remove left spaces       | `" hi".lstrip()` | `"hi"` |
| `rstrip()` | Remove right spaces      | `"hi ".rstrip()` | `"hi"` |

---

### Splitting & Joining

| Function  | Description           | Example               | Output          |
| --------- | --------------------- | --------------------- | --------------- |
| `split()` | Split into list       | `"a,b,c".split(",")`  | `['a','b','c']` |
| `join()`  | Join list into string | `",".join(['a','b'])` | `"a,b"`         |

---

### Partition Functions

| Function       | Description        | Example                        | Output                  |
| -------------- | ------------------ | ------------------------------ | ----------------------- |
| `partition()`  | Split into 3 parts | `"hello=world".partition("=")` | `('hello','=','world')` |
| `rpartition()` | Split from right   | `"a=b=c".rpartition("=")`      | `('a=b','=','c')`       |

---

## Quick Example (Uses Many Functions)

```python
text = "  hello Python 123  "

print(text.strip())              # remove spaces
print(text.upper())              # uppercase
print(text.lower())              # lowercase
print(text.title())              # title case
print(text.swapcase())           # swap case

print(len(text))                 # length
print(text.count("o"))           # count 'o'

print(text.strip().startswith("hello"))
print(text.strip().endswith("123"))

print(text.strip().split())      # split words

words = ["Python", "is", "fun"]
print(" ".join(words))           # join words
```

---

If you want, I can:

* Convert this into a **printable PDF cheat sheet**
* Create **practice questions**
* Make a **mini exam test** based on these functions.
````

````python
# Big Example: Student Data Processing System

-- This program:
1.Takes a student’s name and marks
2.Cleans and formats text
3.Checks different string conditions
4.Uses alignment, searching, splitting, joining, etc.
````

````python
# Student Information Program

# Raw input strings
name = "   vishal pethkar   "
course = "python programming"
marks_str = "85,90,78,92"

print("RAW DATA")
print(name, course, marks_str)
print("-" * 40)

# 1. Strip spaces
name = name.strip()
course = course.strip()

# 2. Case formatting
print("Formatted Name:", name.title())
print("Upper Course:", course.upper())
print("Lower Course:", course.lower())

# 3. Swap case
print("Swapcase Name:", name.swapcase())

# 4. Length, max, min
print("Name Length:", len(name))
print("Max char in name:", max(name))
print("Min char in name:", min(name))

# 5. Check functions
print("Is name alphabetic?", name.replace(" ", "").isalpha())
print("Is course lowercase?", course.islower())

# 6. Searching
print("Count of 'a' in name:", name.count('a'))
print("First index of 'a':", name.find('a'))
print("Last index of 'a':", name.rfind('a'))

# 7. Startswith and endswith
print("Name starts with 'vi'?", name.startswith("vi"))
print("Course ends with 'ing'?", course.endswith("ing"))

# 8. Alignment
print("\nAligned Text")
print(name.center(30, "-"))
print(course.ljust(30, "."))
print(course.rjust(30, "."))

# 9. Splitting marks
marks_list = marks_str.split(",")
print("\nMarks List:", marks_list)

# Convert to integers
marks = []
for m in marks_list:
    marks.append(int(m))

# Calculate total and average
total = sum(marks)
average = total / len(marks)

print("Total Marks:", total)
print("Average Marks:", average)

# 10. Joining words
words = course.split()
joined_course = "-".join(words)
print("Joined Course Name:", joined_course)

# 11. Partition example
email = "student@gmail.com"
user, sep, domain = email.partition("@")
print("\nEmail user:", user)
print("Email domain:", domain)

# 12. Final Report
print("\n" + "=" * 40)
print("STUDENT REPORT".center(40))
print("=" * 40)

print("Name      :", name.title())
print("Course    :", course.title())
print("Marks     :", marks)
print("Total     :", total)
print("Average   :", round(average, 2))

print("=" * 40)
````
