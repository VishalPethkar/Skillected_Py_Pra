```python
#--1.Lambda Function


#What is Lambda?
-- A lambda function is a small anonymous (no-name) function written in one line.
```

```python
#Syntax
lambda arguments: expression
```

```python
#Example 1 – Normal Function
def square(n):
    return n * n

print(square(4))


#Same Using Lambda

square = lambda n: n * n
print(square(4))
```

```python
#Lambda with Two Arguments
add = lambda a, b: a + b
print(add(5, 3))


#Lambda with map()
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, numbers))
print(result)
```

```python
#Short Functions (One Line Function)

#--When function logic is very small, use lambda.

# ❌ Normal Function
def square(n):
    return n * n

print(square(5))


# ✅ Lambda Version
square = lambda n: n * n
print(square(5))

# ✔ Short
# ✔ One line
# ✔ Clean
```

```python
#-- 2.Lambda Inside map()

#What is map()?
map(function, iterable)
#Applies function to every element.
```

```python
#Example: Double Numbers

numbers = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, numbers))
print(result)
```

```python
#-- 3.Lambda Inside filter()

#What is filter()?
filter(function, iterable)
#Keeps elements that return True.
```

```python
#Example: Even Numbers


numbers = [1, 2, 3, 4, 5, 6]
result = list(filter(lambda x: x % 2 == 0, numbers))
print(result)
```

```python
#-- 4.Lambda Inside sorted()


#What is sorted()?
Used to sort data.
```

```python
#Example: Sort by Second Value

students = [("Ritesh", 85), ("Aman", 92), ("Kiran", 78)]
result = sorted(students, key=lambda x: x[1])
print(result)
```

```python
#Real Interview Example

#Sort dictionary by values:

data = {"a": 5, "b": 2, "c": 8}
result = sorted(data.items(), key=lambda x: x[1])
print(result)
```

```python
#REAL DATASET EXAMPLES using:

✔ lambda
✔ map()
✔ filter()
✔ sorted()
```

```python
#Example 1: Student Marks Dataset
students = [
    {"name": "Ritesh", "marks": 85},
    {"name": "Aman", "marks": 92},
    {"name": "Kiran", "marks": 67},
    {"name": "Sneha", "marks": 74}
]
```

```python
#1.Sort Students by Marks

sorted_students = sorted(students, key=lambda x: x["marks"])
print(sorted_students)

# ✔ Used in ranking systems
# ✔ Common interview question
```

```python
#2.Filter Students Above 75

top_students = list(filter(lambda x: x["marks"] > 75, students))
print(top_students)

# ✔ Data filtering
# ✔ Performance analysis
```

```python
#3.Add 5 Grace Marks Using map()


updated_marks = list(map(lambda x: {"name": x["name"], 
                                    "marks": x["marks"] + 5}, students))

print(updated_marks)

# ✔ Data transformation
# ✔ Feature engineering
```

```python
#Example 2: Sales Dataset
sales = [
    {"product": "Laptop", "price": 50000, "quantity": 3},
    {"product": "Phone", "price": 20000, "quantity": 5},
    {"product": "Tablet", "price": 15000, "quantity": 2}
]
```

```python
#1.Calculate Total Revenue Per Product


revenue = list(map(lambda x: x["price"] * x["quantity"], sales))
print(revenue)

# ✔ Business analytics
# ✔ Revenue calculation
```

```python
#2.Sort Products by Revenue


sorted_sales = sorted(sales, 
                      key=lambda x: x["price"] * x["quantity"],
                      reverse=True)

print(sorted_sales)

# ✔ Find top-selling product
```

```python
#Example 3: Employee Salary Dataset
employees = [
    {"name": "Raj", "salary": 40000},
    {"name": "Priya", "salary": 60000},
    {"name": "Anil", "salary": 35000},
    {"name": "Neha", "salary": 70000}
]
```

```python
#1.Filter High Salary Employees (>50000)

high_salary = list(filter(lambda x: x["salary"] > 50000, employees))
print(high_salary)

# ✔ HR analytics
```

```python
#2.Increase Salary by 10%


incremented = list(map(lambda x: {"name": x["name"], 
                                  "salary": x["salary"] * 1.10}, employees))

print(incremented)

# ✔ Payroll system logic
```
