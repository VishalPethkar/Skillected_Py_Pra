```python
#-- LAMBDA CHEATSHEET WITH OUTPUTS
```

```python
#1. Basic Lambda
add = lambda a, b: a + b
print(add(10, 20))
```

```python
#2. Single Parameter
square = lambda x: x * x
print(square(6))
```

```python
#3. Multiple Parameters
avg = lambda a, b, c: (a+b+c)/3
print(avg(10, 20, 30))
```

```python
#4. Conditional (If-Else)
check = lambda x: "Even" if x%2==0 else "Odd"
print(check(9))
```

```python
#5. Lambda with map()
numbers = [1,2,3,4]
result = list(map(lambda x: x*2, numbers))
print(result)
```

```python
#6. Lambda with filter()
numbers = [1,2,3,4,5,6]
even = list(filter(lambda x: x%2==0, numbers))
print(even)
```

```python
#7. Lambda with reduce()
from functools import reduce
numbers = [1,2,3,4]
total = reduce(lambda a,b: a+b, numbers)
print(total)
```

```python
#8. Lambda with sorted()
students = [
    {"name":"Ram","marks":75},
    {"name":"Shyam","marks":60},
    {"name":"Amit","marks":85}
]

sorted_students = sorted(students, key=lambda s: s["marks"])
print(sorted_students)
```

```python
#9. Lambda with max()
numbers = [10, 25, 5, 40]
maximum = max(numbers, key=lambda x: x)
print(maximum)
```

```python
#10. Multi-Condition Lambda
grade = lambda m: "A" if m>=75 else "B" if m>=50 else "C"
print(grade(65))
```

```python
#11. Nested Lambda
add = lambda x: lambda y: x + y
print(add(5)(10))
```

```python
#12. Immediate Execution (IIFE)
print((lambda x: x*x)(7))
```

```python
#13. Sorting Multiple Fields
data = [
    {"name":"A","age":25,"salary":50000},
    {"name":"B","age":25,"salary":30000},
    {"name":"C","age":22,"salary":70000}
]

result = sorted(data, key=lambda x: (x["age"], x["salary"]))
print(result)
```
