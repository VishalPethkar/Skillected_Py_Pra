```python
# 5.What is **kwargs?
**kwargs allows a function to accept any number of keyword arguments
It stores them inside a dictionary
kwargs means → keyword arguments
```

```python
#Basic Example

def student_info(**data):
    print(data)

student_info(name="Ramesh", age=24, city="Pune")
```

```python
#Accessing Dictionary Values

def student_info(**data):
    print("Name:", data["name"])
    print("City:", data["city"])

student_info(name="Ramesh", age=24, city="Pune")
```

```python
#Loop Through **kwargs

def student_info(**data):
    for key, value in data.items():
        print(key, ":", value)

student_info(name="Ramesh", age=24, city="Pune")
```

```python
#Real Industry Example

def employee_record(**details):
    print("Employee Record")
    print("------------------")
    
    for key, value in details.items():
        print(f"{key} : {value}")

employee_record(
    name="Ritesh",
    age=22,
    designation="Data Analyst",
    salary=50000,
    location="Pune"
)
```

```python
# 6.Combine *args and **kwargs

def example(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

example(10, 20, 30, name="Ramesh", city="Pune")
```

```python
#Example

def example02(**kwarg):
    for i, j in zip(kwarg.keys(), kwarg.values()):
        print(i, '->', j)
example02(name='vish', age=21, city='pune')
```

```python
#Better & Professional Way (Recommended)
#Instead of using zip(), use .items() directly.

def example02(**kwarg):
    for key, value in kwarg.items():
        print(key, '->', value)

example02(name='vish', age=21, city='pune')
```

```python
#Real life project



#Now let’s build a Real Project-Level Example using:
✅ Positional Arguments
✅ Default Arguments
✅ *args
✅ **kwargs


#Real Project Example: Employee Payroll Management System

In real companies, payroll systems must handle:
-- Employee basic info
-- Default salary structure
-- Multiple allowances
-- Flexible deductions & extra info
This is exactly where *args and **kwargs are used.
```

```python
#Complete Project-Level Function


def payroll_system(emp_id, emp_name,
                   base_salary=20000,
                   *allowances,
                   **extra_details):

    print("========== PAYROLL SLIP ==========")
    
    # Positional Arguments
    print("Employee ID:", emp_id)
    print("Employee Name:", emp_name)
    
    # Default Argument
    print("Base Salary:", base_salary)
    
    # Variable-Length Arguments (*args)
    total_allowance = sum(allowances)
    print("Allowances:", allowances)
    print("Total Allowances:", total_allowance)
    
    # Keyword Variable-Length Arguments (**kwargs)
    print("\nAdditional Details:")
    total_deduction = 0
    
    for key, value in extra_details.items():
        print(key, ":", value)
        
        # If key contains deduction keyword
        if "deduction" in key.lower():
            total_deduction += value
    
    gross_salary = base_salary + total_allowance
    net_salary = gross_salary - total_deduction
    
    print("\nGross Salary:", gross_salary)
    print("Total Deductions:", total_deduction)
    print("Net Salary:", net_salary)
    print("==================================")
```

```python
#Function Call (Real Company Example)


payroll_system(
    101,                      # emp_id (positional)
    "Vishal",                 # emp_name (positional)
    30000,                    # base_salary (default overridden)
    
    2000, 1500, 1000,         # *allowances
    
    tax_deduction=3000,       # **kwargs
    pf_deduction=2000,
    bonus=5000,
    location="Pune"
)
```

```python
#01.HR Software Example

#In HR systems, employee records must accept flexible data.

def add_employee(emp_id, name, department="General", *skills, **details):
    print("Employee ID:", emp_id)
    print("Name:", name)
    print("Department:", department)
    print("Skills:", skills)
    print("Other Details:", details)

add_employee(
    101,
    "Vishal",
    "IT",
    "Python", "SQL", "Power BI",
    salary=50000,
    location="Pune",
    experience=2
)

# HR software allows unlimited skills & extra details.
```

```python
#02.Banking System Example

#In banking, transactions can include optional metadata.

def process_transaction(account_no, amount,
                        transaction_type="debit",
                        **metadata):

    print("Account No:", account_no)
    print("Amount:", amount)
    print("Transaction Type:", transaction_type)
    print("Metadata:", metadata)

process_transaction(
    12345678,
    5000,
    "credit",
    branch="Pune",
    mode="UPI",
    reference_id="TXN7890"
)

# Banks store flexible transaction details using **kwargs.
```

```python
#03.SaaS Platform Example

#In SaaS apps (like CRM tools), user settings are dynamic.

def create_user(username, role="user", *permissions, **settings):
    print("Username:", username)
    print("Role:", role)
    print("Permissions:", permissions)
    print("Settings:", settings)

create_user(
    "vishal123",
    "admin",
    "read", "write", "delete",
    theme="dark",
    notifications=True,
    region="India"
)

# SaaS systems need flexible permissions & settings.
```

```python
#04.Django & Flask API Example

# Frameworks like
# Django
# Flask
# use *args and **kwargs heavily.

#Example (Flask-style):

def api_response(status, *data, **extra):
    return {
        "status": status,
        "data": data,
        "extra": extra
    }

print(api_response(
    "success",
    {"id": 1, "name": "Vishal"},
    message="User fetched successfully",
    code=200
))

# APIs must return flexible responses.
```
