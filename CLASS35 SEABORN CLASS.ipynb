```python
Correlation (Correct Notes)

Correlation measures how strongly two variables are related to each other.

🔹 Types of Correlation
Positive Correlation (+ve)
When one variable increases, the other also increases
Example: Study time ↑ → Marks ↑


Negative Correlation (-ve)
When one variable increases, the other decreases
Example: Price ↑ → Sales ↓


No Correlation
No relationship between variables
Example: Shoe size and intelligence


Correlation Values (Range)
+1 → Perfect positive correlation
0 → No correlation
-1 → Perfect negative correlation
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns   # ✅ missing import

data = {
    "study_hours": [1,2,3,4,5],
    "marks": [30,40,50,60,70],
    "sleep_hours": [8,7,6,5,4]
}

df = pd.DataFrame(data)
corr = df.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#Pairplot -- In Python (with seaborn), pairplot is used to visualize relationships between all pairs of variables in a dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset("tips")

# Create pairplot
sns.pairplot(df)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset("tips")

# Create pairplot
sns.pairplot(df,hue = "sex")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

sns.pairplot(df, vars=["total_bill", "tip"])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

sns.regplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
sns.pairplot (df,kind = "reg",diag_kind = "kde",hue = "sex")
plt.show()
```

```python
#categorical plot :--  In seaborn, categorical plots are used to visualize relationships where one or both variables are categorical 
#ex -- (like gender, day, class, etc.).
```

```python
#1. Count Plot (frequency count)
sns.countplot(x="day", data=df)
plt.show()
```

```python
#2. Bar Plot (mean values)
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
#3. Swarm Plot (better spacing, no overlap)
sns.swarmplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

plt.figure(figsize=(6,4))

# Barplot (background)
sns.barplot(x="day", y="total_bill", data=df, color="lightblue")

# Swarmplot (on top)
sns.swarmplot(x="day", y="total_bill", data=df, palette="viridis")

plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=df, palette="viridis")
plt.show()
```
