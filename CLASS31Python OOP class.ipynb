```python
#Q.1 create a class with 3 parameters in Python (OOP).

class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

# Creating objects
s1 = Student("Vish", 22, 85)
s2 = Student("Rahul", 23, 90)

# Printing valuessa
print(s1.name, s1.age, s1.marks)
print(s2.name, s2.age, s2.marks)
```

```python
#Q.5  Laptop Class (with show() method) calling method -- 

class Laptop:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def car(self):
        print("Brand:", self.brand)
        print("Price:", self.price)

# Creating objects
l1 = Laptop("HP", 55000)
l2 = Laptop("Dell", 65000)

# Calling method
l1.car()
l2.car()
```

```python
class phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def mob(self):
        print("Brand:", self.brand)
        print("Price:", self.price)


# Creating objects
l1 = phone("iphone", 155000)
l2 = phone("samsung", 165000)

# Calling method
l1.mob()
l2.mob()
```

```python
#method 
#school system
#method = funcation inside a class 




class student:
    def __init__(self,name,mark):

        self.name = name
        self.mark = mark

    def result(self):
        if self.mark >= 50:
            return "nala pass"
        else:
            return "nala fail"

s1 = student("vishu",90)
s2 = student("ritu",0)

print (s1.result())
print (s2.result())
```

```python

#-- inheritance the child class inherits the features of parents class -- 



# Parent Class
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, I am {self.name}")



class Student(Parent):   # Child Class
    pass


class Teacher(Parent):    # Child Class 
    pass


# Objects
s1 = Student("Vishu")
t1 = Teacher("Ritu")

# Calling method
s1.greet()
t1.greet()
```
