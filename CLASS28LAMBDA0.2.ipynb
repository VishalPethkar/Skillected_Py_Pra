```python
#Basic Formula (Syntax)


lambda arguments : expression


#Structure Explanation
-- lambda → Keyword to create anonymous function
-- arguments → Input parameters (like normal function)
-- expression → Single expression (returns value automatically)
-- ⚠️ Lambda function does NOT use return keyword
-- ⚠️ Only one expression is allowed
```

```python
#1. What Happens When You Write Lambda?

#Example:
add = lambda a, b: a + b
```

```python
#What Python Understands:
#It is almost equal to:

def add(a, b):
    return a + b

#-Lambda automatically returns the expression value.
#-No need to write return.
```

```python
#tep-by-Step Working


#Example:
add = lambda a, b: a + b
print(add(10, 20))


#Working Process:

# -lambda a, b: a + b
# -Python creates an anonymous function
# -It stores it in variable add
# -add(10, 20) is called
# -Expression 10 + 20 is calculated
# -Result is returned automatically
```

```python
#2. How Lambda Works with map()

#Example:
numbers = [1, 2, 3]
result = list(map(lambda x: x*2, numbers))
print(result)


#Step-by-Step:
# -map() takes function + iterable
# -Lambda runs on each element
# -Calculation:
# 1 → 1*2 → 2
# 2 → 2*2 → 4
# 3 → 3*2 → 6
# -Returns new list
```

```python
#3. How Lambda Works with filter()


#Example:

numbers = [1,2,3,4]
even = list(filter(lambda x: x%2==0, numbers))
print(even)


#Step-by-Step:
# -filter() checks condition for each value
# -Lambda returns True/False
# -Only True values are kept
# 1 % 2 == 0 → False ❌
# 2 % 2 == 0 → True ✅
# 3 % 2 == 0 → False ❌
# 4 % 2 == 0 → True ✅
```

```python
#4. How Lambda Works with reduce()

#Example:

from functools import reduce

numbers = [1,2,3,4]
total = reduce(lambda a,b: a+b, numbers)
print(total)


#Step-by-Step:
# -Reduce works like this:
# (1 + 2) → 3
# (3 + 3) → 6
# (6 + 4) → 10
```

```python
#Quick Math Operations Using Lambda Function:
```

```python
#1. Addition of Two Numbers
add = lambda a, b: a + b
print(add(5, 3))
```

```python
#2. Square All Numbers in List
numbers = [1, 2, 3, 4]

squares = list(map(lambda x: x * x, numbers))
print(squares)
```

```python
#3. Double Numbers in List
numbers = [1, 2, 3, 4]

double = list(map(lambda x: x * 2, numbers))
print(double)
```

```python
#4. Find Even Numbers
numbers = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```

```python
#5. Sum of List
from functools import reduce

numbers = [1, 2, 3, 4]

total = reduce(lambda a, b: a + b, numbers)
print(total)
```

```python
#6. Find Maximum Number
from functools import reduce

numbers = [5, 10, 3, 8]

maximum = reduce(lambda a, b: a if a > b else b, numbers)
print(maximum)
```

```python
#7. Sorting Short Logic
students = [
    {"name": "Ram", "marks": 80},
    {"name": "Shyam", "marks": 60}
]

sorted_students = sorted(students, key=lambda s: s["marks"])
print(sorted_students)
```

```python
#8. Filtering Simple Condition
numbers = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```

```python
#9. Quick Data Transformation
numbers = [1, 2, 3]

double = list(map(lambda x: x * 2, numbers))
print(double)
```

```python
#10. Simple Conditional Logic
check = lambda x: "Pass" if x >= 40 else "Fail"
print(check(35))
```
