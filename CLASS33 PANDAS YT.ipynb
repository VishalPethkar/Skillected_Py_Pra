```python
#1.groupbys 

import numpy as np
import pandas as pd

data = {
    'category': ["A","B","A","B","A","B","A","B"],
    'store': ["S1","S1","S2","S2","S1","S2","S1","S2"],
    'sale': [100,200,150,250,120,180,200,300],
    'quantity': [10,15,12,18,8,20,15,25],
    'date': pd.date_range("2023-01-01", periods=8)
}

df = pd.DataFrame(data)

print(df)
```

```python
df
```

```python
#qus -- groupby category and calculate the sum of sales 

cat = df.groupby("category")["sale"].sum()
cat
```

```python
#qus -- groupby store and calculate the sum of sales 

cat = df.groupby("store")["sale"].sum()
cat
```

```python
cat = df.groupby('category')      #imp

for i, v in cat:
    print(i)   # group name (A or B)
    print(v)   # group DataFrame
    
```

```python
#groupby multiple columns
#groupby category and store

cat = df.groupby(["category","store"])["sale"].sum()
cat
```

```python
#2.Aggregation -- mutiple values convert - single mean values

#mean , median , min , max , count

df["sale"].agg(["sum", "min", "max", "count"])
```

```python
#Dataframes:--

import pandas as pd
import numpy as np

# Corrected data dictionary
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South','East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice','John', 'Mary', 'Bob', 'Alice','John', 'Mary', 'Bob', 'Alice','John', 'Mary', 'Bob', 'Alice','John', 'Mary', 'Bob', 'Alice']
}

# Create the DataFrame
df = pd.DataFrame(data)


df
```

```python
df["Month"] = df["Date"].dt.month_name()                      #--- add Month
```

```python
df["Quarter"] = 'Q' + df["Date"].dt.quarter.astype(str)           #--- add Quarter values
```

```python
df
```

```python
#1.Pivot tables: --- using a aggregation par
#pivot table is used to summarize data into a strucated table format

#df.pivot_table (values = "col",index = "col",aggfunc = "func")


#values -- calculate --- index,row,aggfunc
```

```python
#creating pivot taable:--

pd.pivot_table(df,values = "Sales",index = "Region",columns = "Product")
```

```python
#creating pivot taable:-- use aggrgation fun (to find median valus)

pd.pivot_table(df,values = "Sales",index = "Region",columns = "Product",aggfunc = "median")
```

```python
pivot2 = pd.pivot_table(df,values = ["Sales","Units"],index = "Region",columns = "Product")
print(pivot2)
```
