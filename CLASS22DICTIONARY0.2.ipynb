```python
️#1.get()
# Purpose: Gets value of a key safely (no error if key not found).
```

```python
d = {"name": "Vishal", "age": 21}

print(d.get("name"))        # Vishal
print(d.get("salary"))      # None
print(d.get("salary", 0))   # 0


# ✔ Prevents KeyError
# ✔ Can give default value
```

```python
#2.keys()
# Purpose: Returns all keys.

print(d.keys())
```

```python
#3.values()
# Purpose: Returns all values.

print(d.values())
```

```python
#4.items()
# Purpose:Returns key + value pairs (Best for loops)

for k, v in d.items():
    print(k, "->", v)
```

```python
#5.len()
# Purpose:Counts number of key-value pairs.

print(len(d))
```

```python
#6.copy()
# Purpose:Creates copy of dictionary.

d2 = d.copy()
print(d2)

#✔ Changes in d2 won’t affect original d
```

```python
#7.fromkeys()
# Purpose: Creates new dictionary with given keys.

keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)
print(new_dict)
```

```python
#8.setdefault()
# Purpose:Returns value if key exists.
#        If not → adds key with default value.

d.setdefault("salary", 50000)
print(d)
```

```python
#9.update()
# Purpose: Updates dictionary with another dictionary.

d.update({"city": "Mumbai"})
print(d)

# ✔ Adds new key
# ✔ Updates existing key
```

```python
#10.pop(key, default)
# Purpose:Removes specific key.

d.pop("age")
print(d)
```

```python
#11.popitem()
# Purpose:Removes last inserted key-value pair.

d.popitem()
print(d)

#⚠ Works LIFO (Last In First Out)
```

```python
#EXAMPLES -- Real-Life Example: Employee Database System

# Think of a company office system.
# Each employee has:
# Employee ID

--Name
--Age
--Department
--Salary

In real life:
👉 Employee ID → unique
👉 Employee details → connected to that ID

That is exactly what a dictionary does.
```

```python
employees = {
    101: {"name": "Vishal", "age": 21, "dept": "IT", "salary": 40000},
    102: {"name": "Rahul", "age": 25, "dept": "HR", "salary": 35000},
    103: {"name": "Amit", "age": 28, "dept": "Finance", "salary": 50000}
}
```

```python
#Accessing Employee Details
print(employees[101])
```

```python
#Get Only Name
print(employees[101]['name'])
```

```python
#Add New Employee
employees[104] = {"name": "Neha", "age": 23, "dept": "Marketing", "salary": 30000}
print(employees)
```

```python
#Update Salary
employees[101]['salary'] = 45000
print(employees)
```

```python
#Remove employee
employees.pop(102)
print(employees)
```

```python
#Loop Through All Employees
for emp_id, details in employees.items():
    print(emp_id, "->", details['name'])
```

```python
#-- real life mini project -- 


#Student Management System (Mini Project)
-This project can:
-Add student
-View student
-Update marks
-Delete student
-Show all students
```

```python
# Step 1: Data Structure

# Each student has:
# Roll number (unique key)
# Name
# Age
# Marks

students = {}
```

```python
students = {}

while True:
    print("\n--- Student Management System ---")
    print("1. Add Student")
    print("2. View Student")
    print("3. Update Marks")
    print("4. Delete Student")
    print("5. Show All Students")
    print("6. Exit")

    choice = input("Enter choice: ")

    # 1️⃣ Add Student
    if choice == "1":
        roll = int(input("Enter Roll Number: "))
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        marks = int(input("Enter Marks: "))

        students[roll] = {
            "name": name,
            "age": age,
            "marks": marks
        }

        print("Student Added Successfully!")

    # 2️⃣ View Student
    elif choice == "2":
        roll = int(input("Enter Roll Number: "))
        if roll in students:
            print(students[roll])
        else:
            print("Student Not Found!")

    # 3️⃣ Update Marks
    elif choice == "3":
        roll = int(input("Enter Roll Number: "))
        if roll in students:
            new_marks = int(input("Enter New Marks: "))
            students[roll]["marks"] = new_marks
            print("Marks Updated!")
        else:
            print("Student Not Found!")

    # 4️⃣ Delete Student
    elif choice == "4":
        roll = int(input("Enter Roll Number: "))
        if roll in students:
            students.pop(roll)
            print("Student Deleted!")
        else:
            print("Student Not Found!")

    # 5️⃣ Show All Students
    elif choice == "5":
        for roll, details in students.items():
            print(roll, "->", details)

    # 6️⃣ Exit
    elif choice == "6":
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice!")
```
