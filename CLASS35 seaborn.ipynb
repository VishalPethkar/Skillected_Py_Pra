```python
import seaborn as sns
import matplotlib.pyplot as plt

# style
sns.set_style("darkgrid")
sns.set_context("talk")

# load data
df = sns.load_dataset("tips")

# plot
sns.scatterplot(
    x="total_bill",
    y="tip",
    data=df,
    hue="day",
    size="size",
    style="sex"
)

plt.title("Bill vs Tips Analysis")
plt.show()
```
