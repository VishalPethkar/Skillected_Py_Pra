```python
#-- What is a Docstring?

-- A docstring (documentation string) is a string written inside triple quotes (""" """ or ''' ''') that explains what a function, class, or module does.
-- It is used for documentation.
```

```python
#Why Docstring is Important?

✔ Explains purpose of code
✔ Makes code readable
✔ Used in large projects
✔ Helps in documentation tools
✔ Used in interviews & companies
```

```python
#Basic Syntax

def function_name():
    """This is a docstring"""
    pass
```

```python
#Example 1: Simple Function Docstring
def add(a, b):
    """This function returns the sum of two numbers"""
    return a + b
```

```python
#To read docstring:

print(add.__doc__)
```

```python
#OR

help(add)
```

```python
#2.Class Docstring
class Student:
    """This class represents a student"""

    def __init__(self, name):
        self.name = name
```

```python
#3.Module Docstring

#At the top of file:

"""
This module performs basic math operations.
"""

def add(a, b):
    return a + b
```

```python
#-- Multi-Line Docstring (Professional Style)


def divide(a, b):
    """
    This function divides two numbers.

    Parameters:
    a (int/float): Numerator
    b (int/float): Denominator

    Returns:
    float: Result of division
    """
    return a / b

#This is industry standard format.
```

```python
#Single Line vs Multi-Line

#--1.Single Line (Short)
def square(n):
    """Returns square of number"""
    return n*n
```

```python
#-- 2.Multi-Line (Detailed)

def square(n):
    """
    Returns square of number.

    Parameter:
    n (int): Input number

    Returns:
    int: Square of n
    """
    return n*n
```

```python
#--EXAMPLE--


#1.Real Life Example: Student Result Analyzer

#With Docstring (Professional Way)


def calculate_average(marks):
    """
    Calculates the average of student marks.

    Parameters:
    marks (list): A list of numerical marks.

    Returns:
    float: The average value of marks.
    """
    return sum(marks) / len(marks)
```

```python
#2.Real Business Example: Salary Increment System


def apply_increment(salary, percentage):
    """
    Applies percentage increment to employee salary.

    Parameters:
    salary (float): Current salary of employee.
    percentage (float): Increment percentage.

    Returns:
    float: Updated salary after increment.
    """
    return salary + (salary * percentage / 100)
```

```python
#3.Real Banking Example


def withdraw(balance, amount):
    """
    Withdraws amount from account balance.

    Parameters:
    balance (float): Current account balance.
    amount (float): Amount to withdraw.

    Returns:
    float: Updated balance after withdrawal.
    """
    if amount > balance:
        return "Insufficient Balance"
    return balance - amount
```

```python
def strrev(str1):
    """
    Reverses the given string and prints original and reversed string.

    Parameters:
    str1 (str): Input string

    Returns:
    None
    """

    print(f"Original string: {str1}")
    rev = str1[::-1]
    print(f"Reversed string: {rev}")


strrev('vish peth')
```
