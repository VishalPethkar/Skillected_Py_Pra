````python
# -- COPY FUN
#1.SHALLOW COPY
#2.DEEP COPY
````

````python
#1.SHALLOW COPY

#- A shallow copy creates a new object
#- But nested objects are still shared (same memory)




#- Example:
import copy

a = [[1, 2], [3, 4]]
b = copy.copy(a)   # shallow copy

b[0][0] = 100

print("Original:", a)
print("Copy:", b)
````

````python
#2.DEEP COPY

#- A deep copy creates a new object
#- AND copies all nested objects also
#- Completely independent copy



#- Example:
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)   # deep copy

b[0][0] = 100

print("Original:", a)
print("Copy:", b)
````

````python
import copy

# Original Data (Nested List)
students = [
    {
        "name": "Rahul",
        "marks": [80, 75, 90]
    },
    {
        "name": "Sneha",
        "marks": [85, 88, 92]
    }
]

print("===== ORIGINAL DATA =====")
print(students)

# ----------------------------
# SHALLOW COPY
# ----------------------------
shallow_students = copy.copy(students)

# Modify inner data
shallow_students[0]["marks"][0] = 100

print("\n===== AFTER SHALLOW COPY CHANGE =====")
print("Original:", students)
print("Shallow :", shallow_students)


# ----------------------------
# DEEP COPY
# ----------------------------
deep_students = copy.deepcopy(students)

# Modify inner data
deep_students[1]["marks"][1] = 50

print("\n===== AFTER DEEP COPY CHANGE =====")
print("Original:", students)
print("Deep    :", deep_students)
````

````python
#- EXAPLIN I DEATIL:


#  WHY COPY IS IMPORTANT?

In Python, variables do NOT store actual data.
They store **reference (memory address)** of the object.

Example:

```python
a = [1, 2, 3]
b = a
```

Here:

* `a` and `b` point to SAME memory.
* Changing `b` will change `a`.

That’s why we need **copying**.

---

# 1️⃣ SHALLOW COPY (Detailed Explanation)

## 📌 Definition

A **Shallow Copy**:

* Creates a new outer object
* BUT inner objects are still shared (same memory reference)

---

## 📌 Example with Memory Concept

```python
import copy

a = [[10, 20], [30, 40]]
b = copy.copy(a)

b[0][0] = 999
```

---

## 🧠 Memory Structure

Before change:

```
a  ──►  [  list1 , list2  ]
b  ──►  [  list1 , list2  ]  (new outer list)

list1 ──► [10, 20]
list2 ──► [30, 40]
```

Notice:

* Outer list is different
* But inner lists are SAME

After:

```
b[0][0] = 999
```

Both become:

```
[[999, 20], [30, 40]]
```

Because inner list is shared.

---

## 📌 When Shallow Copy is Safe?

✅ When there are no nested objects
Example:

```python
a = [1, 2, 3]
b = a.copy()
```

Works perfectly fine.

---

# 2️⃣ DEEP COPY (Detailed Explanation)

## 📌 Definition

A **Deep Copy**:

* Creates new outer object
* Creates new inner objects also
* Fully independent

---

## 📌 Example

```python
import copy

a = [[10, 20], [30, 40]]
b = copy.deepcopy(a)

b[0][0] = 999
```

---

## 🧠 Memory Structure

Before change:

```
a ──► [ list1 , list2 ]
b ──► [ new_list1 , new_list2 ]

list1 ──► [10, 20]
new_list1 ──► [10, 20]
```

Everything is separate.

After change:

```
a = [[10, 20], [30, 40]]
b = [[999, 20], [30, 40]]
```

Original remains unchanged ✅

---

# 🔥 BIG REAL-LIFE EXAMPLE (Interview Level)

Imagine:
You have student database:

```python
students = [
    {"name": "Vishal", "marks": [85, 90]},
    {"name": "Riya", "marks": [78, 88]}
]
```

Now:

If you use shallow copy and update marks,
👉 Original database also changes 😱
This can cause data corruption in:

* Machine Learning training data
* Backend APIs
* Banking systems
* E-commerce cart systems

So in real projects → **Deep copy is safer for nested data**

---

# ⚡ Difference Summary (Clear Concept)

| Feature         | Shallow Copy | Deep Copy |
| --------------- | ------------ | --------- |
| Outer object    | New          | New       |
| Inner object    | Same         | New       |
| Memory Shared   | Yes          | No        |
| Speed           | Faster       | Slower    |
| Safe for nested | ❌ No         | ✅ Yes     |

---

# 🎯 Important Interview Questions

1. What is the difference between copy() and deepcopy()?
2. Why does shallow copy fail with nested lists?
3. Is tuple affected in shallow copy?
4. Which copy is memory expensive?
5. When should deep copy be avoided?

---

# 🧠 Advanced Concept (Very Important)

Shallow copy copies:

* Object structure
* But NOT recursive content

Deep copy:

* Recursively copies everything

---

# 💡 Final Rule (Remember This)

👉 No nested structure → Shallow copy OK
👉 Nested structure present → Use Deep copy


````
