```python
#Import:
import numpy as np
```

```python
#Why NumPy?
✔ Faster than Python lists
✔ Uses less memory
✔ Supports vectorized operations
✔ Useful for matrices & scientific computing
```

```python
#-- NumPy Array --
```

```python
#1. Creating Array


import numpy as np

arr = np.array([1, 2, 3, 4])
print(arr)
```

```python
#2. Multi-Dimensional Array

arr2 = np.array([[1, 2], [3, 4]])
print(arr2)
```

```python
#3. Check Dimensions

print(arr.ndim)   # dimension
print(arr.shape)  # size
```

```python
#--Special Arrays

np.zeros((2,3))   # all 0
np.ones((2,3))    # all 1
np.eye(3)         # identity matrix
np.arange(0,10,2) # range
np.linspace(0,1,5) # equal parts
```
