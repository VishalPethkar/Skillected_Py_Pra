```python
#PYTHON - 05 - Python Operators
# # Operators are symbols used to perform operations on variables and values.
# # Types of Python Operators



# ### 1. Arithmetic Operators
# Used for mathematical operations.
# | Operator | Meaning             | Example  | Result |
# | -------- | ------------------- | -------- | ------ |
# | `+`      | Addition            | `5 + 3`  | 8      |
# | `-`      | Subtraction         | `5 - 3`  | 2      |
# | `*`      | Multiplication      | `5 * 3`  | 15     |
# | `/`      | Division            | `6 / 3`  | 2.0    |
# | `//`     | Floor division      | `7 // 3` | 2      |
# | `%`      | Modulus (remainder) | `7 % 3`  | 1      |
# | `**`     | Power               | `2 ** 3` | 8      |



    
# Example:
# a = 7
# b = 3
# print(a + b)
# print(a % b)



# ### 2. Comparison (Relational) Operators
# Used to compare values. Result is **True** or **False**.
# | Operator | Meaning          | Example  |
# | -------- | ---------------- | -------- |
# | `==`     | Equal to         | `5 == 5` |
# | `!=`     | Not equal to     | `5 != 3` |
# | `>`      | Greater than     | `6 > 2`  |
# | `<`      | Less than        | `2 < 6`  |
# | `>=`     | Greater or equal | `5 >= 5` |
# | `<=`     | Less or equal    | `3 <= 5` |




# ### 3. Logical Operators
# Used to combine conditions.
# | Operator | Meaning        | Example          |
# | -------- | -------------- | ---------------- |
# | `and`    | Both true      | `True and False` |
# | `or`     | Any one true   | `True or False`  |
# | `not`    | Reverse result | `not True`       |

    
# Example:
# a = 10
# print(a > 5 and a < 20)



# ### 4. Assignment Operators
# Used to assign values.
# | Operator | Example  | Same as     |
# | -------- | -------- | ----------- |
# | `=`      | `x = 5`  | assign      |
# | `+=`     | `x += 3` | `x = x + 3` |
# | `-=`     | `x -= 2` | `x = x - 2` |
# | `*=`     | `x *= 4` | `x = x * 4` |
# | `/=`     | `x /= 2` | `x = x / 2` |



    

# ### 5. Bitwise Operators
# Used for binary operations.
# | Operator | Meaning     |    |
# | -------- | ----------- | -- |
# | `&`      | AND         |    |
# | `        | `           | OR |
# | `^`      | XOR         |    |
# | `~`      | NOT         |    |
# | `<<`     | Left shift  |    |
# | `>>`     | Right shift |    |


    

# ### 6. Membership Operators

# Check if value exists in a sequence.
# | Operator | Example              |
# | -------- | -------------------- |
# | `in`     | `'a' in "apple"`     |
# | `not in` | `'x' not in "apple"` |



# ### 7. Identity Operators

# Check if two variables refer to the same object.

# | Operator | Example      |
# | -------- | ------------ |
# | `is`     | `a is b`     |
# | `is not` | `a is not b` |



# Example
# a = 5
# b = 2
# print(a + b)      # Arithmetic
# print(a > b)      # Comparison
# print(a > 3 and b < 5)   # Logical

```

```python
#EXAMPLES

## 1. Arithmetic Operators

a = 10
b = 3
print(a + b)   # Addition → 13
print(a - b)   # Subtraction → 7
print(a * b)   # Multiplication → 30
print(a / b)   # Division → 3.333...
print(a // b)  # Floor division → 3
print(a % b)   # Modulus → 1
print(a ** b)  # Power → 1000




## 2. Comparison Operators

a = 5
b = 10
print(a == b)   # False
print(a != b)   # True
print(a > b)    # False
print(a < b)    # True
print(a >= 5)   # True
print(b <= 10)  # True




## 3. Logical Operators

x = 8
print(x > 5 and x < 10)   # True
print(x > 10 or x < 5)    # False
print(not(x > 5))         # False




## 4. Assignment Operators

x = 5
x += 3   # x = x + 3
print(x) # 8

x *= 2   # x = x * 2
print(x) # 16




## 5. Membership Operators

text = "python"
print("p" in text)      # True
print("z" not in text)  # True




## 6. Identity Operators

a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)      # True (same object)
print(a is c)      # False (different objects)
print(a is not c)  # True




## 7. Bitwise Operators

a = 5   # 101 in binary
b = 3   # 011 in binary

print(a & b)  # AND → 1
print(a | b)  # OR  → 7
print(a ^ b)  # XOR → 6


#7.1 -- & -- AND
0 & 1
0 & 0 
1 & 1
0 & 0


#7.2 -- / -- OR
0 | 1
0 | 0 
1 | 1
0 | 0


#7.3 -- ^ -- XOR
0 ^ 1
0 ^ 0 
1 ^ 1
0 ^ 0


#7.4 -- ~ -- COMPLEMENT
a = 2 
print (bin(a))
print (~a)
print (bin(~a))


#7.5 -- << -- LEFT SHIFT
#1.
a = 20
print(bin(a))
print(a<<2)


#7.6 -- << -- RIGHT SHIFT
#2.
a = 20
print(bin(a))
print(a>>2)
```

```python
#8.PRINT ADDITION VALUES (even and odd count)

i = 1
odd_sum = 0
even_sum = 0

while i <= 10:
    a = int(input('Enter any number: '))
    
    if a % 2 == 0:
        even_sum += a
    else:
        odd_sum += a       
    
    i += 1

print('Even sum:', even_sum)
print('Odd sum:', odd_sum)

```

```python
#9.PRINT COUNT OF POSITIVE AND NEGATIVE VALUES

i = 1
Pos = 0
Neg = 0

while i <= 10:
    a = int(input('Enter any number: '))
    
    if a > 0:
        Pos += 1
    elif a < 0:
        Neg += 1
    else:
        print('Zero entered')
    
    i += 1

print('Positive count:', Pos)
print('Negative count:', Neg)
```

```python
#10. Program to find maximum and minimum numbers

n = int(input("Enter how many numbers you want to enter: "))

num = int(input("Enter number: "))
maximum = num
minimum = num

for i in range(1, n):
    num = int(input("Enter number: "))
    
    if num > maximum:
        maximum = num
    elif num < minimum:
        minimum = num

print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

```python
# PRINT ODD AND EVEN
# PRINT POSITIVE AND NEGATIVE
# PRINT MINIMUM AND MAXIMUM
# PRINT TOTAL NUMBERS ENTERED

odd = 0
even = 0
positive = 0
negative = 0
count = 0

minimum = None
maximum = None

print("Enter numbers (enter 0 to stop):")

while True:
    num = int(input())

    if num == 0:
        break

    count += 1

    # Odd / Even
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

    # Positive / Negative
    if num > 0:
        positive += 1
    else:
        negative += 1

    # Minimum
    if minimum is None or num < minimum:
        minimum = num

    # Maximum
    if maximum is None or num > maximum:
        maximum = num

# Output
print("\nTotal numbers entered:", count)
print("Odd numbers:", odd)
print("Even numbers:", even)
print("Positive numbers:", positive)
print("Negative numbers:", negative)
print("Minimum number:", minimum)
print("Maximum number:", maximum)
```
