```python
##Step 1: Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Step 2: Create Database

# Create Student Database

data = {
    "Student": ["Amit", "Rahul", "Priya", "Sneha", "Vikas", "Neha", "Rohan", "Pooja", "Kiran", "Anjali"],
    "Age": [20, 21, 19, 22, 20, 21, 23, 20, 22, 19],
    "Maths": [78, 85, 90, 67, 88, 76, 95, 80, 72, 89],
    "Science": [82, 79, 91, 70, 86, 75, 93, 84, 74, 90],
    "English": [74, 88, 85, 60, 80, 79, 92, 77, 70, 87],
    "Gender": ["Male", "Male", "Female", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "City": ["Pune", "Mumbai", "Delhi", "Pune", "Delhi", "Mumbai", "Pune", "Delhi", "Mumbai", "Pune"]
}

# Create DataFrame

df = pd.DataFrame(data)

print(df)
```

```python
## Step 3: Apply Important Functions
# Basic Functions

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.isnull().sum())
```

```python
#Statistical Functions

print(df["Maths"].mean())
print(df["Science"].median())
print(df["English"].mode())
print(df["Maths"].max())
print(df["Science"].min())
print(df[["Maths", "Science", "English"]].corr())
```

```python
#Sorting Functions

print(df.sort_values(by="Maths", ascending=False))
```

```python
#Filtering Data

print(df[df["Maths"] > 80])
```

```python
#Groupby Function

print(df.groupby("Gender")[["Maths", "Science"]].mean())
```

```python
##Step 4: Matplotlib Plots

# 1. Line Plot

plt.figure(figsize=(8,5))
plt.plot(df["Student"], df["Maths"], marker='o')
plt.title("Student Maths Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#2. Bar Plot

plt.figure(figsize=(8,5))
plt.bar(df["Student"], df["Science"])
plt.title("Science Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#3. Histogram

plt.figure(figsize=(8,5))
plt.hist(df["Maths"], bins=5)
plt.title("Maths Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#4. Scatter Plot

plt.figure(figsize=(8,5))
plt.scatter(df["Maths"], df["Science"])
plt.title("Maths vs Science")
plt.xlabel("Maths")
plt.ylabel("Science")
plt.show()
```

```python
#5. Pie Chart

city_count = df["City"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(city_count, labels=city_count.index, autopct='%1.1f%%')
plt.title("Students by City")
plt.show()
```

```python
#6. Box Plot

plt.figure(figsize=(8,5))
plt.boxplot(df[["Maths", "Science", "English"]])
plt.xticks([1,2,3], ["Maths", "Science", "English"])
plt.title("Box Plot")
plt.show()
```

```python
##Step 5: Seaborn Plots

#1. Heatmap

corr = df[["Maths", "Science", "English"]].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#2. Pairplot

sns.pairplot(df, hue="Gender")
plt.show()
```

```python
#3. Regplot

plt.figure(figsize=(7,5))
sns.regplot(x="Maths", y="Science", data=df)
plt.title("Regression Plot")
plt.show()
```

```python
##Step 6: Distribution Plots

#1. Histplot

plt.figure(figsize=(7,5))
sns.histplot(df["Maths"], kde=True)
plt.title("Histplot")
plt.show()
```

```python
#2. KDE Plot

plt.figure(figsize=(7,5))
sns.kdeplot(df["Science"], fill=True)
plt.title("KDE Plot")
plt.show()
```

```python
#3. Rug Plot

plt.figure(figsize=(7,5))
sns.rugplot(df["English"])
plt.title("Rug Plot")
plt.show()
```

```python
#4. Joint Plot

sns.jointplot(x="Maths", y="Science", data=df, kind="scatter")
plt.show()
```

```python
#5. Distplot (Old Version)

sns.distplot(df["Maths"])
plt.show()
```

```python
##Step 7: Categorical Data Plots

# 1. Count Plot

plt.figure(figsize=(7,5))
sns.countplot(x="Gender", data=df)
plt.title("Count Plot")
plt.show()
```

```python
#2. Bar Plot

plt.figure(figsize=(7,5))
sns.barplot(x="Gender", y="Maths", data=df)
plt.title("Bar Plot")
plt.show()
```

```python
#3. Box Plot

plt.figure(figsize=(7,5))
sns.boxplot(x="Gender", y="Science", data=df)
plt.title("Box Plot")
plt.show()
```

```python
#4. Violin Plot

plt.figure(figsize=(7,5))
sns.violinplot(x="Gender", y="English", data=df)
plt.title("Violin Plot")
plt.show()
```

```python
#5. Strip Plot

plt.figure(figsize=(7,5))
sns.stripplot(x="Gender", y="Maths", data=df)
plt.title("Strip Plot")
plt.show()
```

```python
#6. Swarm Plot

plt.figure(figsize=(7,5))
sns.swarmplot(x="Gender", y="Science", data=df)
plt.title("Swarm Plot")
plt.show()
```

```python
#7. Point Plot

plt.figure(figsize=(7,5))
sns.pointplot(x="Gender", y="English", data=df)
plt.title("Point Plot")
plt.show()
```

```python
##Step 8: Combine Two Plots


#1.Barplot + Swarmplot
plt.figure(figsize=(8,5))
sns.barplot(x="Gender", y="Maths", data=df)
sns.swarmplot(x="Gender", y="Maths", data=df)
plt.title("Combined Plot")
plt.show()
```

```python
##Step 9: Color Palettes

sns.palplot(sns.color_palette("deep"))
plt.show()

sns.palplot(sns.color_palette("pastel"))
plt.show()

sns.palplot(sns.color_palette("dark"))
plt.show()
```

```python
##Step 10: Save Plot

plt.figure(figsize=(7,5))
sns.barplot(x="Gender", y="Maths", data=df)
plt.title("Save Plot Example")
plt.savefig("myplot.png")
plt.show()
```
