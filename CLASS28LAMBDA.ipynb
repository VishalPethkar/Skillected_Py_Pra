```python
#What is a Lambda Function?
--A Lambda Function in Python is a small anonymous (nameless) function created using the lambda keyword.

It is used when:
-- You need a short function
-- The function is used only once
-- You don’t want to define a full function using def
-- Fun can have multiple arguments but only one expression which is evaluated and returned.
-- we can have a lambda definition anywhere in a program where a funcation is expected it can or cannot be assigned to a variable.
```

```python
#Syntax


lambda arguments : expression

# lambda → keyword
# arguments → input parameters
# expression → single expression (only one line allowed)
# ⚠️ Important: Lambda can contain only one expression, not multiple statements.
```

```python
# Limitations of Lambda

-- Cannot use multiple expressions
-- Cannot include loops or complex statements
-- Less readable if logic is complex
```

```python
#Example
```

```python
#Normal Function

def add(a, b):
    return a + b

print(add(10, 20))
```

```python
#Lambda Function

add = lambda a, b: a + b
print(add(10, 20))
```

```python
#Lambda with Single Argument

square = lambda x: x * x
print(square(5))
```

```python
#Lambda with No Argument

greet = lambda: "Hello Python"
print(greet())
```

```python
#Lambda with Multiple Arguments

max_num = lambda a, b: a if a > b else b
print(max_num(10, 20))
```

```python
#Lambda with Built-in Functions (Very Important)

Lambda is mostly used with:

-- map()
-- filter()
-- reduce()
-- sorted()
```

```python
#1. Lambda with map()

#Used to apply function to every element.

numbers = [1, 2, 3, 4]

square = list(map(lambda x: x*x, numbers))
print(square)
```

```python
#2. Lambda with filter()

#Used to filter elements based on condition.

numbers = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```

```python
#3. Lambda with reduce()
#reduce() is inside functools module.

from functools import reduce
numbers = [1, 2, 3, 4]
sum_all = reduce(lambda a, b: a + b, numbers)
print(sum_all)
```

```python
#4. Lambda with sorted()


students = [("Ram", 23), ("Shyam", 21), ("Amit", 25)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
```

```python
#-- Mini Project Using Lambda in Python


#Employee Management System (Console-Based)
--This mini project uses lambda functions with:
- sorted()
- filter()
- map()
- reduce()
It simulates a small HR Software system where you manage employees.

 

#Project Features
✅ Add Employee
✅ View All Employees
✅ Sort by Salary
✅ Filter High Salary Employees
✅ Increase Salary (Using map + lambda)
✅ Calculate Total Salary (Using reduce + lambda)
```

```python
#Step 1: Complete Code
from functools import reduce

# Employee Data
employees = [
    {"id": 1, "name": "Vishal", "salary": 50000},
    {"id": 2, "name": "Rahul", "salary": 30000},
    {"id": 3, "name": "Amit", "salary": 70000},
]

def add_employee():
    id = int(input("Enter ID: "))
    name = input("Enter Name: ")
    salary = float(input("Enter Salary: "))
    
    employees.append({"id": id, "name": name, "salary": salary})
    print("Employee Added Successfully!\n")

def view_employees():
    for emp in employees:
        print(emp)
    print()

def sort_by_salary():
    sorted_emp = sorted(employees, key=lambda x: x["salary"])
    print("Sorted by Salary:")
    for emp in sorted_emp:
        print(emp)
    print()

def high_salary_filter():
    high_salary = list(filter(lambda x: x["salary"] > 50000, employees))
    print("High Salary Employees (>50000):")
    for emp in high_salary:
        print(emp)
    print()

def increase_salary():
    increased = list(map(lambda x: {"id": x["id"],
                                    "name": x["name"],
                                    "salary": x["salary"] * 1.10}, employees))
    print("Salary Increased by 10%:")
    for emp in increased:
        print(emp)
    print()

def total_salary():
    total = reduce(lambda a, b: a + b["salary"], employees, 0)
    print("Total Salary of All Employees:", total)
    print()

# Menu
while True:
    print("1. Add Employee")
    print("2. View Employees")
    print("3. Sort by Salary")
    print("4. Filter High Salary")
    print("5. Increase Salary by 10%")
    print("6. Total Salary")
    print("7. Exit")
    
    choice = int(input("Enter Choice: "))
    
    if choice == 1:
        add_employee()
    elif choice == 2:
        view_employees()
    elif choice == 3:
        sort_by_salary()
    elif choice == 4:
        high_salary_filter()
    elif choice == 5:
        increase_salary()
    elif choice == 6:
        total_salary()
    elif choice == 7:
        print("Thank You!")
        break
    else:
        print("Invalid Choice\n")
```

```python
#Where Lambda is Used?

Feature                      	Lambda Usage
Sorting	                        key=lambda x: x["salary"]
Filtering	                    lambda x: x["salary"] > 50000
Mapping	                        lambda x: modified dictionary
Reducing	                    lambda a, b: a + b["salary"]
```

```python
#5. keyword varibles length

e = lambda ** kwargs:kwargs
e(one = 1,two = 2,three = 3)
```

```python
#Example 1: Return All Keyword Arguments
e1 = lambda **kwargs: kwargs

print(e1(one=1, two=2, three=3))
```

```python
#Example 2: Get Only Values
e1 = lambda **kwargs: list(kwargs.values())

print(e1(one=1, two=2, three=3))
```

```python
#Example 3: Get Only Keys
e1 = lambda **kwargs: list(kwargs.keys())

print(e1(one=1, two=2, three=3))
```

```python
#Example 4: Sum of Values
e1 = lambda **kwargs: sum(kwargs.values())

print(e1(one=1, two=2, three=3))
```
