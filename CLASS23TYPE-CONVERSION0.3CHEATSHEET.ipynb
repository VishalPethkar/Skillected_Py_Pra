```python
#1.Algebra
✔ (a + b)²

(a+b)^2 = a^2 + 2ab + b^2
122=64+32+32+16=144
```

```python
Example:
If a = 3, b = 2

= 3² + 2(3×2) + 2²
= 9 + 12 + 4
= 25
```

```python
✔ (a − b)²

(a-b)^2 = a^2 - 2ab + b^2
2=64−24−24+9=25
```

```python
Example:
If a = 5, b = 1

= 25 − 10 + 1
= 16
```

```python
✔ Difference of Squares

a^2 - b^2 = (a-b)(a+b)
```

```python
Example:
9² − 4²
= (9−4)(9+4)
= 5 × 13
= 65
```

```python
#3.Statistics
✔ Mean

Mean = Σx / n

Example:
Numbers: 10, 20, 30

Mean = (10+20+30)/3
= 60/3
= 20
```

```python
✔ Variance


σ2=Σ(x−μ)2/N

Example:
Data: 2, 4, 6
Mean = 4

Variance = [(2−4)² + (4−4)² + (6−4)²] / 3
= (4 + 0 + 4)/3
= 2.67
```

```python
#1.Quadratic Formula: x=(−b±√(b2−4ac))/(2a)

import math

a = 1
b = -3
c = 2

d = b*b - 4*a*c
root1 = (-b + math.sqrt(d)) / (2*a)
root2 = (-b - math.sqrt(d)) / (2*a)

print("Roots:", root1, root2)
```

```python
#2.Arithmetic Progression (AP):an​=a+(n−1)d
#nth Term

a = 2
d = 3
n = 5

an = a + (n-1)*d
print("Nth Term:", an)
```

```python
#Sum of First n Terms:

a = 2
d = 3
n = 5

Sn = n/2 * (2*a + (n-1)*d)
print("Sum:", Sn)
```
