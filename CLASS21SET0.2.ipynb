```python
#intersection_update()

s1 = {1,2,3,4,5,6}
s2 = {7,8,9,1,2,3}

print(s1)
print(s2)
print(s1.intersection_update(s2))
print(s1)
print(s2)
```

```python
#intersection()

s1 = {1,2,3,4,5,6}
s2 = {7,8,9,1,2,3}

print(s1)
print(s2)

print(s1.intersection(s2))
print(s1)
print(s2)

print(s1&s2)                                            #and
print(s1|s2)                                            #or

print(s1)
```
