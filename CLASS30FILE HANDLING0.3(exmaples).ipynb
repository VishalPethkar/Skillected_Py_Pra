```python
#1. Read File (Basic)
with open("data.txt", "r") as f:
    print(f.read())
```

```python
#2. Write File (Overwrite)
with open("data.txt", "w") as f:
    f.write("Hello Python\n")
    f.write("File Handling Example")
```

```python
#3. Append File (Add Data)
with open("data.txt", "a") as f:
    f.write("\nNew Line Added")
```

```python
#4. Read Line by Line
with open("data.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
#5. Using readlines()
with open("data.txt", "r") as f:
    data = f.readlines()
    print(data)
```

```python
#6. Remove \n (Clean Output)
with open("data.txt", "r") as f:
    for line in f.readlines():
        print(line.strip())
```

```python
#7. Read First N Characters
with open("data.txt", "r") as f:
    print(f.read(5))
```

```python
#8. File Pointer (seek)
with open("data.txt", "r") as f:
    print(f.read(5))
    f.seek(0)
    print(f.read(5))
```

```python
#9. Check File Exists (Using Exception)
try:
    with open("data.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found")
```

```python
#10. Write Multiple Records
names = ["Rahul", "Om", "Vishal"]

with open("students.txt", "w") as f:
    for name in names:
        f.write(name + "\n")
```

```python
# --- MINI PROJECT: STUDENT RECORD SYSTEM

# -- Features
# ✔ Add student
# ✔ View students
# ✔ Search student
```

```python
def add_student():
    name = input("Enter name: ")
    age = input("Enter age: ")
    course = input("Enter course: ")

    with open("students.txt", "a") as f:
        f.write(name + "-" + age + "-" + course + "\n")

    print("Student added successfully!")


def view_students():
    try:
        with open("students.txt", "r") as f:
            for line in f:
                print(line.strip())
    except FileNotFoundError:
        print("No records found!")


def search_student():
    search = input("Enter name to search: ")
    found = False

    try:
        with open("students.txt", "r") as f:
            for line in f:
                if search.lower() in line.lower():
                    print("Found:", line.strip())
                    found = True
    except FileNotFoundError:
        print("File not found")

    if not found:
        print("Student not found")


# Menu
while True:
    print("\n1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        print("Thank you!")
        break
    else:
        print("Invalid choice")
```
