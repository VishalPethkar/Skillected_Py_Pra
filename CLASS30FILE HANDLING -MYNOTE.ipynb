```python
#File Handling in Python 

-- File handling in Python allows you to create, read, write, and update files. 
It is used to store data permanently instead of keeping it only in program memory.
    
-- Python provides built-in functions like open(), read(), write(), close(), etc.
```

```python
#1.--Opening a File

#Syntax:

file = open("filename", "mode")
```

```python
#Example:

file = open("data.txt", "r")
print(file.read())
file.close()



#Parameters
# Parameter	 -- Meaning
# filename	 -- Name of the file
# mode	     -- How the file should be opened
```

```python
#2. --File Modes
# Mode	  Meaning
# "r"	      Read file
# "w"	      Write file (creates new or overwrite)
# "a"	      Append data
# "x"	      Create new file
# "b"    	  Binary mode
# "t"	      Text mode (default)
```

```python
#Example:

file = open("data.txt", "w")
file.write("Hello Python")
file.close()
```

```python
#3. --Using with open() (Best Method)

# This automatically closes the file.

with open("data.txt", "r") as file:
    print(file.read())

# Advantages:
# ✔ No need to use close()
# ✔ Cleaner code
# ✔ Prevents memory issues
```

```python
#4.--Reading Files
#1. read() – Read whole file

with open("data.txt", "r") as f:
    data = f.read()
    print(data) 
```

```python
#2. read(n) – Read specific characters

with open("data.txt", "r") as f:
    print(f.read(5))
```

```python
#3. readline() – Read one line

with open("data.txt", "r") as f:
    print(f.readline())
```

```python
#4. readlines() – Read all lines into a list

with open("data.txt", "r") as f:
    data = f.readlines()
    print(data)
```

```python
#5.-- Removing \n (New Line)

#Since you recently asked about this in your file handling practice, here is the correct method.


with open("data.txt", "r") as f:
    data = f.readlines()

for line in data:
    print(line.strip())
```

```python
line.rstrip()   # remove right spaces
line.lstrip()   # remove left spaces
line.strip()    # remove both sides
```

```python
#6. --Writing to a File


#Example
with open("data.txt", "w") as f:
    f.write("This is Python\n")
    f.write("File Handling Example\n")

#"w" overwrites existing content.
```

```python
#7. --Append Data to File


with open("data.txt", "a") as f:
    f.write("New line added\n")
```

```python
#8. --Writing User Input to File


name = input("Enter name: ")
age = input("Enter age: ")

with open("students.txt", "a") as f:
    f.write(name + "-" + age + "\n")
```

```python
#9. --File Pointer & seek()

# File pointer shows current position in file.

# Example:

with open("data.txt", "r") as f:
    print(f.read(5))
    f.seek(0)
    print(f.read(5))


#seek(0) moves pointer back to start.
```

```python
#10. --Check if File Exists

import os

if os.path.exists("data.txt"):
    print("File exists")
else:
    print("File not found")
```

```python
#11. --Delete a File

import os
os.remove("data.txt")
```

```python
#12. --Practical Example (Student Record System)


name = input("Enter Name: ")
age = input("Enter Age: ")
course = input("Enter Course: ")

with open("students.txt", "a") as f:
    f.write(name + "-" + age + "-" + course + "\n")

with open("students.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
#Summary:--


# Function	     Purpose
# open()	     Open file
# read()	     Read file
# readline()	 Read one line
# readlines()    Read all lines
# write()	     Write data
# append()	     Add data
# seek()	     Move pointer
# close()	     Close file
```

```python
#Mini Project: Student Record System Using File Handling (Python)

# This mini project stores student details in a text file and allows you to:

# 1️.Add Student
# 2️.View Students
# 3️.Search Student
# 4️.Delete Student
# 5️.Exit Program
```

```python
import os

file_name = "students.txt"

def add_student():
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    course = input("Enter Course: ")

    with open(file_name, "a") as f:
        f.write(name + "-" + age + "-" + course + "\n")

    print("Student Added Successfully\n")


def view_students():
    if os.path.exists(file_name):
        with open(file_name, "r") as f:
            data = f.readlines()

        if len(data) == 0:
            print("No records found\n")
        else:
            print("\nStudent Records:")
            for line in data:
                print(line.strip())
    else:
        print("File not found\n")


def search_student():
    name = input("Enter student name to search: ")

    found = False
    with open(file_name, "r") as f:
        for line in f:
            if name.lower() in line.lower():
                print("Record Found:", line.strip())
                found = True

    if not found:
        print("Student not found\n")


def delete_student():
    name = input("Enter student name to delete: ")

    lines = []
    found = False

    with open(file_name, "r") as f:
        lines = f.readlines()

    with open(file_name, "w") as f:
        for line in lines:
            if name.lower() not in line.lower():
                f.write(line)
            else:
                found = True

    if found:
        print("Student Deleted Successfully\n")
    else:
        print("Student not found\n")


while True:

    print("===== Student Record System =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_students()

    elif choice == "3":
        search_student()

    elif choice == "4":
        delete_student()

    elif choice == "5":
        print("Program Closed")
        break

    else:
        print("Invalid choice\n")
```
