```python
#7-- 5 Types of Arguments (Very Important 🔥)
```

```python
#1.Positional Arguments

#Arguments are passed in correct order.

def add(a, b):
    return a + b

print(add(5, 10))

#Order matters.
```

```python
#2.Keyword Arguments

#Arguments passed using parameter names.

print(add(b=10, a=5))

#Order does NOT matter.
```

```python
#3.Default Arguments

# We assign default value to parameter.

def greet(name="User"):
    print("Hello", name)

greet()
greet("Ritesh")

# If no argument passed → default value used.
```

```python
#4.Variable-Length Arguments (*args)

# Used when we don’t know how many arguments.

def total(*numbers):
    return sum(numbers)

print(total(10, 20, 30))

# *args stores values in tuple.
```

```python
#5.Keyword Variable-Length Arguments (**kwargs)

#Used for multiple keyword arguments.

def details(**info):
    print(info)

details(name="Ritesh", age=22)

#**kwargs stores values in dictionary.
```

```python
#MINI PROJECT: Bank Management System (Using Functions)

👉 Concepts Used:
--Functions
--Arguments
--Return
--Conditional statements
--Loop
--Basic logic

🎯 Project Features

1️⃣ Create Account
2️⃣ Deposit Money
3️⃣ Withdraw Money
4️⃣ Check Balance
5️⃣ Exit
```

```python
balance = 0

def create_account():
    global balance
    balance = 0
    print("Account created successfully!")

def deposit(amount):
    global balance
    balance += amount
    print("Amount deposited successfully!")

def withdraw(amount):
    global balance
    if amount > balance:
        print("Insufficient balance!")
    else:
        balance -= amount
        print("Withdrawal successful!")

def check_balance():
    print("Current Balance:", balance)

def menu():
    print("\n--- Bank Menu ---")
    print("1. Create Account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Check Balance")
    print("5. Exit")

while True:
    menu()
    choice = int(input("Enter your choice: "))

    if choice == 1:
        create_account()

    elif choice == 2:
        amount = float(input("Enter deposit amount: "))
        deposit(amount)

    elif choice == 3:
        amount = float(input("Enter withdrawal amount: "))
        withdraw(amount)

    elif choice == 4:
        check_balance()

    elif choice == 5:
        print("Thank you for using our bank!")
        break

    else:
        print("Invalid choice! Try again.")
```
