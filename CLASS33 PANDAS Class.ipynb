```python
import pandas as pd

data = {
    "Name": ["Amit","Riya","Rahul","Sneha","Karan","Pooja","Arjun","Neha"],
    "City": ["Pune","Mumbai","Delhi","Pune","Mumbai","Delhi","Pune","Mumbai"],
    "Department": ["IT","HR","IT","Finance","HR","IT","Finance","HR"],
    "Salary": [50000,60000,55000,52000,58000,62000,51000,59000]
}

df = pd.DataFrame(data)
df
```
