```python
#-- TUPLE IN PYTHON (Complete Guide)



#What is a Tuple?

- Ordered collection
- Immutable (cannot change after creation)
- Allows duplicate values
- Can store multiple data types
- A tuple is similar to a list fun
- A tuple is create by placing all the elements inside brackets() separated by commas
- A tuple can have any number of item and they may or may not be different data types
```

```python
[start : stop : step]
step + ve
start < stop
step + ve
start > stop
```

```python
#- EXAMPLE

l1 = [1,2,3,4,5]
print(l1[-1])
print(l1[::-1])
```

```python
#-- creating a empty tuple

a = ()
print(a)
print(type(a))
```

```python
#-- single value tuple

a = (1,)
print(a)
print(type(a))
```

```python
#-- multiple value tuple

a = 1,2,3.14,'anil','true',2+4j
print(a)
print(type(a))
```

```python
t = (5,)
print(type(t))   # tuple ✅
```

```python
t = (5)
print(type(t))   # int ❌
```

```python
#-- Accessing Tuple Elements
t = (10, 20, 30, 40)

print(t[0])      # 10
print(t[-1])     # 40
print(t[1:3])    # (20, 30)
```

```python
#-- Looping in Tuple
t = (5, 10, 15)

for i in t:
    print(i)
```

```python
#Tuple Functions


t = (10, 20, 30, 20)

print(len(t))        # Length
print(max(t))        # Maximum
print(min(t))        # Minimum
print(t.count(20))   # Count
print(t.index(30))   # Index position
```

```python
#-- unpacking and packing

#What is Packing?
- Packing means putting multiple values into a single variable (usually a tuple).
```

```python
#Example 1: Simple Packing


a = 10
b = 20
c = 30

t = (a, b, c)   # Packing
print(t)
```

```python
#Example 2: Automatic Packing (Very Important 🔥)


t = 1, 2, 3   # No brackets needed
print(t)
print(type(t))
```

```python
#-- What is Unpacking?
- Unpacking means extracting values from a tuple into separate variables.
```

```python
#Example 1: Basic Unpacking


t = (100, 200, 300)

x, y, z = t   # Unpacking

print(x)
print(y)
print(z)
```

```python
#Advanced Unpacking (Using *)

This is very important for interviews 
```

```python
t = (1, 2, 3, 4, 5)

a, *b, c = t

print(a)   # 1
print(b)   # [2, 3, 4]
print(c)   # 5
```

```python
| Function | Work                 |
| -------- | -------------------- |
| len()    | Count elements       |
| max()    | Largest value        |
| min()    | Smallest value       |
| sum()    | Total of numbers     |
| tuple()  | Convert to tuple     |
| count()  | Count specific value |
| index()  | Find position        |
```

```python
t = (10, 20, 30, 20, 50)
```

```python
#1.len() → Length of Tuple
#Definition:Returns total number of elements.

t = (10, 20, 30, 20, 50)
print(len(t))
```

```python
#2.max() → Maximum Value
#Definition:Returns largest element.

print(max(t))
```

```python
#3.min() → Minimum Value
#Definition: Returns smallest element.

print(min(t))
```

```python
#4.sum() → Total of Elements
#Definition: Adds all numeric elements.

print(sum(t))
```

```python
#5.tuple() → Convert into Tuple
#Definition: Converts iterable (list, string, set) into tuple.

#Example 1: List → Tuple
l = [1, 2, 3]
t2 = tuple(l)
print(t2)
```

```python
#6.t.count() → Count Occurrence
#Definition: Returns how many times value appears.

print(t.count(20))
```

```python
#7.t.index() → Find Position
#Definition: Returns index of first occurrence.

print(t.index(30))
```

```python
t = (10, 20, 30, 20, 50)

print("Length:", len(t))
print("Maximum:", max(t))
print("Minimum:", min(t))
print("Sum:", sum(t))
print("Count of 20:", t.count(20))
print("Index of 30:", t.index(30))
```
