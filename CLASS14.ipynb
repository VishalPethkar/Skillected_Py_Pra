```python
# PYTHON-- 14 -- FOR LOOP


#A.--for loop -- It is used to iterate over a sequence (string,list,tuple,etc).

#syntax
for var in sequence:                  # var is a dummy varible defind locally for the loop
    statements
else:                                 # "else" is optional
    statements





#B.--range() -- Fun is used to as an iterator.
#it return a sequence of num,string from 0 by default and increate by 1 by default and stop before a specified num 

#syntax 
range(start,stop,step)
```

```python
#EXAMPLE

range(7)
range(0,7)
for i in range (0,7):
    print(1)
```

```python
#EXAMPLE

range(7)
range(0,7)
for i in range (0,7):
    print(i)
```

```python
#EXAMPLE
#PRINT A NUM FROM 1 TO 100 ALL VALUES

#.1 -- for loop --
for i in range (1,101):
    print (i,end = ' ')
    
```

```python
#.2 -- while loop -- 
for i in range (1,101):
    print (i,end = ' ')
    i += 1
```

```python
#.3 -- PRINT ODD NUM --

for i in range (1,101,2):
    print (i,end = ' ')
    i += 1
```

```python
#.3 -- PRINT EVEN NUM --

for i in range (0,101,2):
    print (i,end = ' ')
    i += 1
```
