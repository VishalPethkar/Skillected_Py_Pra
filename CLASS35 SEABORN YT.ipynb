```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)
y = x ** 2

fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x, y)
axis1.set_title("Main Graph")

axis2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
axis2.plot(x, y)
axis2.set_title("Small Graph")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)

fig, axes = plt.subplots(1, 2)

# First plot
axes[0].plot(x, x**2)
axes[0].set_title("Square")

# Second plot
axes[1].plot(x, x**3)
axes[1].set_title("Cube")

plt.show()
```

```python
## -- Matrix Plots in Seaborn -- Matrix plots are used to visualize data in the form of a 2D matrix using colors.

#They help in:
- Finding patterns
- Detecting correlation
- Identifying missing values
- Understanding large datasets visually
```

```python
#Types of Matrix Plots in Seaborn
Plot	        Use
heatmap()	    Correlation & matrix visualization
clustermap()	Group similar rows/columns
```

```python
#Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1. Heatmap -- A heatmap uses colors to represent values in a matrix.

# Syntax
sns.heatmap(data)
```

```python
#Example 1 : Simple Heatmap
import numpy as np

data = np.random.rand(5,5)

sns.heatmap(data)

plt.show()
```

```python
#Example 2 : Correlation Heatmap
tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only=True)

sns.heatmap(corr)

plt.show()
```

```python
#Example 3 : Heatmap with Values
sns.heatmap(corr,
            annot=True)

plt.show()


# Meaning -- annot=True → shows values inside boxes
```

```python
#Example 4 : Change Color Map
sns.heatmap(corr,
            annot=True,
            cmap="coolwarm")

plt.show()
```

```python
#Common Colormaps
Colormap	Description
coolwarm	Blue to red
viridis	    Green-yellow-purple
magma	    Dark to light
plasma	    Purple-yellow
YlGnBu	    Yellow-green-blue
```

```python
#Example 5 : Line Widths
sns.heatmap(corr,
            annot=True,
            linewidths=2)

plt.show()
```

```python
#Example 6 : Change Figure Size
plt.figure(figsize=(8,6))

sns.heatmap(corr,
            annot=True)

plt.show()
```

```python
#Example 7 : Remove Color Bar
sns.heatmap(corr,
            cbar=False)

plt.show()
```

```python
#Example 8 : Show Only Upper Triangle
import numpy as np

mask = np.triu(corr)

sns.heatmap(corr,
            mask=mask,
            annot=True)

plt.show()
```

```python
# Important Parameters in Heatmap
# Parameter	Use
# annot=True	Show values
# cmap=""	    Change colors
# linewidths	Border size
# cbar=False	Remove side color bar
# mask	        Hide some values
# fmt=".2f"	    Decimal formatting
```

```python
# 2. Clustermap -- Groups similar rows and columns together.

-- Useful in:
Data analysis
Pattern recognition
Machine learning preprocessing


#Syntax
sns.clustermap(data)
```

```python
#Example 1 : Simple Clustermap
sns.clustermap(corr)

plt.show()
```

```python
#Example 2 : Clustermap with Colors
sns.clustermap(corr,
               cmap="viridis",
               annot=True)

plt.show()
```

```python
#Example 3 : Standard Scaling
sns.clustermap(corr,
               standard_scale=1)

plt.show()

# Meaning -- standard_scale=1 → normalize columns
```

```python
#Difference Between Heatmap and Clustermap
#Heatmap	                    Clustermap
Simple matrix display   	Groups similar data
No clustering           	Hierarchical clustering
Faster                  	More analytical
Easy to read             	Better for patterns
```

```python
#BIG EXAMPLE -- FULL

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only=True)

plt.figure(figsize=(8,6))

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            linewidths=1)

plt.title("Correlation Heatmap")

plt.show()
```
