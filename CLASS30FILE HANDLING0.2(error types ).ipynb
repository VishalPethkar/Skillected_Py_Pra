```python
# -- File Handling Errors -- 

#What are File Handling Errors?
-File handling errors are exceptions raised when something goes wrong while working with files 
    (open, read, write, close).
    -- In Python, these errors are part of the Exception Handling system.
```

```python
#-- TOPIC 1:FILE HANDLING --
1. Definition:-- File Handling in Python is used to store, read, and manage data permanently in files like 
    .txt, .csv, etc.
```

```python
2. Key Concepts

-- Use open() to work with files
-- Different modes: read, write, append
-- Use with (best practice)
-- Used in Data Science to handle datasets
```

```python
3. Syntax
# Basic
file = open("file.txt", "r")

# Recommended
with open("file.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
4. File Modes
Mode	Meaning
r	Read
w	Write (overwrite)
a	Append
r+	Read + Write
```

```python
5. Important Points

# -- 'w' deletes old data 
# -- 'a' adds data at end
# -- read() → full content
# -- readlines() → list of lines
# -- write() → no newline (use \n)
# -- File not found → FileNotFoundError
```

```python
6. Example
# Write
with open("student.txt", "w") as f:
    f.write("Name: Rahul\n")

# Read
with open("student.txt", "r") as f:
    print(f.read())
```

```python
#TOPIC 2: EXCEPTION HANDLING

 1. Definition:--  Exception Handling is used to handle errors so the program doesn’t crash.
```

```python
2. Keywords

-- try → risky code
-- except → handle error
-- else → runs if no error
-- finally → always runs
```

```python
3. Syntax
try:
    x = int(input("Enter number: "))
except ValueError:
    print("Invalid input")
else:
    print("No error")
finally:
    print("Done")
```

```python
4. Common Exceptions


-- Error	              Meaning
- ValueError	          Wrong value
- TypeError	          Wrong type
- ZeroDivisionError	  Divide by 0
- IndexError	          List index error
- KeyError	          Dict key missing
- FileNotFoundError	  File missing
```

```python
# 1. ValueError → Wrong Value
# Occurs when the value is invalid for the operation.

x = int("abc")   # ❌ cannot convert string to int
```

```python
#✔ Example:

try:
    x = int("abc")
except ValueError:
    print("Invalid value")
```

```python
# 2. TypeError → Wrong Type
# Occurs when operation is applied on wrong data type.

x = "10" + 5   # ❌ string + integer
```

```python
#✔ Example:

try:
    x = "10" + 5
except TypeError:
    print("Type mismatch")
```

```python
#3. ZeroDivisionError → Divide by 0
# Occurs when a number is divided by zero.

x = 10 / 0   # ❌ not allowed
```

```python
#✔ Example:

try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
#4. IndexError → List Index Error
# Occurs when accessing invalid index in list.

a = [1, 2, 3]
print(a[5])   # ❌ index out of range
```

```python
#✔ Example:

try:
    a = [1,2,3]
    print(a[5])
except IndexError:
    print("Index out of range")
```

```python
#5. KeyError → Dictionary Key Missing
# Occurs when accessing a key that does not exist.

d = {"name": "Rahul"}
print(d["age"])   # ❌ key not found
```

```python
#✔ Example:

try:
    d = {"name": "Rahul"}
    print(d["age"])
except KeyError:
    print("Key not found")
```

```python
#6. FileNotFoundError → File Missing
# Occurs when trying to open a file that does not exist.

open("data.txt", "r")   # ❌ file not found
```

```python
#✔ Example:

try:
    open("data.txt", "r")
except FileNotFoundError:
    print("File not found")
```

```python
5. Example


try:
    num = int(input("Enter number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
finally:
    print("Program finished")
```

```python
#-- ERROR BLOCK --
```

```python
# Definition:-- Error block is used to handle runtime errors using:
👉 try, except, else, finally
```

```python
#Structure

try:
    # risky code
except:
    # handle error
else:
    # runs if no error
finally:
    # always runs
```

```python
#Block Meaning

try → code that may cause error
except → runs if error occurs
else → runs if NO error
finally → always runs
```

```python
#-- MULTIPLE EXAMPLES -- 
```

```python
#1. Basic Example
try:
    x = int("abc")
except ValueError:
    print("Invalid input")
```

```python
#2. Multiple Except Blocks
try:
    x = int(input("Enter number: "))
    print(10 / x)
except ValueError:
    print("Wrong value")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
#3. Using else Block
try:
    x = int("10")
except ValueError:
    print("Error")
else:
    print("No error")
```

```python
#4. Using finally Block
try:
    x = 10 / 2
except:
    print("Error")
finally:
    print("This always runs")
```

```python
#5. Full Example (All Blocks)
try:
    x = int(input("Enter number: "))
    result = 10 / x
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Result:", result)
finally:
    print("Program finished")
```

```python
#6. Catch Error Message (as e)
try:
    x = int("abc")
except ValueError as e:
    print("Error:", e)
```

```python
#7. File Handling with Exception
try:
    with open("data.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found")
```

```python
#8. Multiple Errors in One Block
try:
    x = int(input("Enter: "))
    print(10 / x)
except (ValueError, ZeroDivisionError):
    print("Error occurred")
```

```python
#EXAM SHORT POINTS – EXCEPTION HANDLING

✔ try → Contains risky code (may cause error)
✔ except → Handles the error if it occurs
✔ else → Runs only when NO error occurs
✔ finally → Always runs (error or no error)
✔ Use specific exceptions → Better debugging (e.g., ValueError, TypeError)
```
