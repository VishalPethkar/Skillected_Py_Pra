```python
#SET in Python – Complete Detailed Information


#What is a Set?
- A set in Python is:
- An unordered collection of elements
- Mutable (can be changed)
- Does NOT allow duplicate values
- Written using curly braces {}
```

```python
my_set = {1, 2, 3, 4}
```

```python
#Characteristics of Set

   Feature	                                         Description
- Unordered	                              No indexing (no order like list)
- Unique values	                          Duplicate values automatically removed
- Mutable	You can                       add/remove elements
- No Indexing	                          Cannot access using index like list
- Fast operations	                      Useful for mathematical operations
```

```python
#Creating a Set:


1️.Using Curly Braces
s = {1, 2, 3}


️2.Using set() Constructor
s = set([1, 2, 3])


️3.Empty Set
s = set()      # correct
s = {}         # WRONG (creates dictionary)
```

```python
#-- EXAMPLE -- All Set Methods in Python (With Explanation & Example)
```

```python
#1.add()

#Adds one element to the set.

s = {1, 2}
s.add(3)
print(s)
```

```python
#2.update()

#Adds multiple elements (list, tuple, set).

s = {1, 2}
s.update([3, 4])
print(s)
```

```python
#3.remove()

#Removes element.
#❌ Gives error if element not found.

s = {1, 2, 3}
s.remove(2)
```

```python
#4.discard()

# Removes element.
# ✅ No error if element not found.

s = {1, 2, 3}
s.discard(5)
```

```python
#5.pop()

#Removes random element (since set is unordered).

s = {1, 2, 3}
s.pop()
```

```python
#6.clear()

#Removes all elements.

s = {1, 2, 3}
s.clear()
```

```python
#7.copy()

#Returns a copy of set.

s1 = {1, 2}
s2 = s1.copy()
```

```python
#8.union()

#Combines two sets.

a = {1, 2}
b = {2, 3}
print(a.union(b))
```

```python
#9.intersection()

#Common elements.

a = {1, 2, 3}
b = {2, 3, 4}
print(a.intersection(b))
```

```python
#10.difference()

#Elements in first set but not in second.

a = {1, 2, 3}
b = {2, 4}
print(a.difference(b))
```

```python
#11.symmetric_difference()

#Elements in both sets except common ones.

a = {1, 2, 3}
b = {2, 4}
print(a.symmetric_difference(b))
```

```python
#12.intersection_update()

#Updates set with common elements.

a = {1, 2, 3}
b = {2, 3, 4}
a.intersection_update(b)
```

```python
#13.difference_update()

#Updates set with difference.

a = {1, 2, 3}
b = {2}
a.difference_update(b)
```

```python
#-- Set Function in Python Dictionary (Detailed Explanation)

In Python, set() and dictionary (dict) are different data types — but we can use set() with dictionaries in many ways.
```

```python
#1.What is Dictionary?

#A dictionary stores data in key : value pairs.

student = {
    "name": "Ravi",
    "age": 21,
    "course": "B.Tech"
}

print(student)
```

```python
#2.Using set() with Dictionary

#When you apply set() to a dictionary, it returns only the keys.

student = {
    "name": "Ravi",
    "age": 21,
    "course": "B.Tech"
}

print(set(student))
```

```python
#3.Get Keys as Set
student = {
    "name": "Ravi",
    "age": 21,
    "course": "B.Tech"
}

keys_set = set(student.keys())
print(keys_set)
```

```python
#4.Get Values as Set

#Removes duplicate values automatically.
student = {
    "name": "Ravi",
    "age": 21,
    "course": "B.Tech"
}

values_set = set(student.values())
print(values_set)
```

```python
#5.Get Items as Set (Key, Value Pairs)
student = {
    "name": "Ravi",
    "age": 21
}

items_set = set(student.items())
print(items_set)
```

```python
#6.Remove Duplicate Values from Dictionary
data = {
    "a": 10,
    "b": 20,
    "c": 10
}

unique_values = set(data.values())
print(unique_values)
```

```python
#7.Find Common Keys Between Two Dictionaries
dict1 = {"a": 1, "b": 2, "c": 3}
dict2 = {"b": 20, "c": 30, "d": 40}

common_keys = set(dict1.keys()).intersection(dict2.keys())
print("Common Keys:", common_keys)
```

```python
#8.Find Common Values Between Two Dictionaries
dict1 = {"a": 1, "b": 2, "c": 3}
dict2 = {"x": 2, "y": 3, "z": 5}

common_values = set(dict1.values()).intersection(dict2.values())
print("Common Values:", common_values)
```

```python
#-- Dictionary vs Set Difference
Feature	                   Set	                   Dictionary
Syntax	                 {1,2,3}	              {"a":1}
Stores	               Only values	              Key : Value
Duplicate	           Not allowed	              Keys not allowed duplicate
Ordered              No (before 3.7)	          Yes (Python 3.7+)
```

```python
#-- Real-Life Example
#Find students who are present in both classes

classA = {"Ravi": 101, "Amit": 102, "Suresh": 103}
classB = {"Amit": 201, "Suresh": 202, "Karan": 203}

common_students = set(classA.keys()).intersection(classB.keys())

print("Common Students:", common_students)
```

```python
#Important Points

✔ set(dict) returns only keys
✔ Use dict.keys() for keys
✔ Use dict.values() for values
✔ Use dict.items() for key-value pairs
✔ Set operations help compare dictionaries
```

```python
#-- Mini Project: Student Management System


#Objective:
-- Manage student records using a Python dictionary.
We will use:
- add student
- update student
- delete student
- search student
- display all students


#Concept Used:
- Dictionary
- Loop
- If-Else
- Functions
- Input
```

```python
students = {}

def add_student():
    roll = input("Enter Roll No: ")
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    course = input("Enter Course: ")

    students[roll] = {
        "Name": name,
        "Age": age,
        "Course": course
    }
    print("Student Added Successfully!\n")


def display_students():
    if not students:
        print("No student records found.\n")
    else:
        for roll, details in students.items():
            print("Roll No:", roll)
            for key, value in details.items():
                print(f"{key}: {value}")
            print("------------------")


def search_student():
    roll = input("Enter Roll No to Search: ")
    if roll in students:
        print("Student Found:")
        for key, value in students[roll].items():
            print(f"{key}: {value}")
    else:
        print("Student Not Found!")


def delete_student():
    roll = input("Enter Roll No to Delete: ")
    if roll in students:
        del students[roll]
        print("Student Deleted Successfully!")
    else:
        print("Student Not Found!")


while True:
    print("\n---- Student Management System ----")
    print("1. Add Student")
    print("2. Display Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        display_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        delete_student()
    elif choice == "5":
        print("Exiting Program...")
        break
    else:
        print("Invalid Choice! Try Again.")
```
