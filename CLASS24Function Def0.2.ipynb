```python
#FUN DEFINITION

def example():
    print("I am vishal the men!")
```

```python
#FUNCATION DEFINITION -- WITH PARAMETERS

def employee(name):
    print("I am employee of the year!")
    print(f"employee name:{name} ")
```

```python
employee('vish')
```

```python
employee(123456)
```

```python
#CREATE A DYNAMIC

a = input('enter your name :')
employee(a)
```

```python
#FUNCATION ARGUMENT ARE PASSED AS A REFERENCE

employee ([1,2,3])

#STORGE LIST
```

```python
#EX

def change(a1):
    print(f"fun before any changes list = {a1}")
    a1 [-1] = 100
    print(f"fun after any changes list = {a1}")

```

```python
a1 = [1,2,3,4]
print(f"fun before any changes list = {a1}")
```

```python
change(a1)
```

```python
print(f"fun after any changes list = {a1}")
```
