```python
# 1.Fibonacci Series
fibonacci.py
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

        # 
# 2️.Prime Number
prime.py
def is_prime(num):
    if num <= 1:
        return "Not Prime"

    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"

    return "Prime"

    
# ️3.Palindrome Number
palindrome.py
def palindrome(num):
    temp = num
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    if temp == rev:
        return "Palindrome"
    else:
        return "Not Palindrome"

        
# ️4.Armstrong Number
armstrong.py
def armstrong(num):
    temp = num
    sum = 0

    while num > 0:
        digit = num % 10
        sum = sum + digit ** 3
        num = num // 10

    if sum == temp:
        return "Armstrong Number"
    else:
        return "Not Armstrong"

        
# ️5.Even / Odd
even_odd.py
def even_odd(num):
    if num % 2 == 0:
        return "Even Number"
    else:
        return "Odd Number"

        
# ️6.GCD
gcd.py
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

```python
#main.py (Run All Programs)


from logic.fibonacci import fibonacci
from logic.prime import is_prime
from logic.palindrome import palindrome
from logic.armstrong import armstrong
from logic.even_odd import even_odd
from logic.gcd import gcd

print("Fibonacci Series:")
fibonacci(10)

print("\nPrime Check:", is_prime(7))

print("Palindrome Check:", palindrome(121))

print("Armstrong Check:", armstrong(153))

print("Even/Odd:", even_odd(10))

print("GCD:", gcd(12, 18))
```
