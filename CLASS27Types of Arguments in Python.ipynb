```python
#-- Types of Arguments in Python
```

```python
#There are 5 main types of arguments in Python:

-Positional Arguments
-Keyword Arguments
-Default Arguments
-Variable-Length Arguments (*args)
-Keyword Variable-Length Arguments (**kwargs)
```

```python
#1.Positional Arguments

#Arguments are passed in the same order as parameters in the function definition.

#Example:
def greet(name, age):
    print("Name:", name)
    print("Age:", age)

greet("Ritesh", 22)
```

```python
#Order is important:

greet(22, "Ritesh")  # Wrong order
```

```python
#2.Keyword Arguments

#Arguments are passed using parameter name = value format.


#Example:
def greet(name, age):
    print("Name:", name)
    print("Age:", age)

greet(age=22, name="Ritesh")
```

```python
#3.Default Arguments

#A parameter can have a default value.
#If no value is provided during function call, default value is used.

#Example:
def greet(name, age=18):
    print("Name:", name)
    print("Age:", age)

greet("Ritesh")
```

```python
#f we pass age:

greet("Ritesh", 22)
```

```python
#4.Variable-Length Arguments (*args)

# Used when we don’t know how many arguments will be passed.
# It collects values in a tuple.

# Example:
def add_numbers(*args):
    print(args)

add_numbers(1, 2, 3, 4)
```

```python
#Real Example:

def total(*numbers):
    print(sum(numbers))

total(10, 20, 30)
```

```python
#5.Keyword Variable-Length Arguments (**kwargs)

# Used to accept multiple keyword arguments.
# It collects them in a dictionary.

# Example:
def student_info(**kwargs):
    print(kwargs)

student_info(name="Ritesh", age=22, city="Pune")
```

```python
#Important Rules
#Order of Arguments in Function Definition:


def function_name(positional, default, *args, **kwargs):

Correct Order:
-- Positional
-- Default
-- *args
-- **kwargs
```

```python
#Example Using All Types


def example(a, b=10, *args, **kwargs):
    print("a:", a)
    print("b:", b)
    print("args:", args)
    print("kwargs:", kwargs)

example(5, 20, 30, 40, name="Ritesh", city="Pune")
```

```python
#Summary Table
Type	                  Description	             Data Type
Positional	              Based on position	         Normal value
Keyword	                  name=value format	         Normal value
Default	                  Has default value	         Normal value
*args	                  Multiple values	         Tuple
**kwargs	              Multiple keyword values	 Dictionary
```

```python
# BIG Example: Student Admission System

# This function uses:
-- Positional Argument
-- Default Argument
-- *args
-- **kwargs
```

```python
def student_admission(name, age, course="B.Tech", *skills, **details):
    
    print("----- STUDENT ADMISSION DETAILS -----")
    
    # Positional Arguments
    print("Name:", name)
    print("Age:", age)
    
    # Default Argument
    print("Course:", course)
    
    # Variable-Length Arguments (*args)
    print("Skills:", skills)
    
    # Keyword Variable-Length Arguments (**kwargs)
    print("Additional Details:")
    for key, value in details.items():
        print(key, ":", value)
```

```python
#Function Call

student_admission(
    "Ritesh",              # Positional
    22,                    # Positional
    "Computer Science",    # Default overridden
    "Python", "SQL", "ML", # *args
    city="Pune",           # **kwargs
    percentage=77.30,
    internship="YBI Foundation"
)
```

```python
#-- EXAMPLES -- 
#Another Big Practical Example (Salary Calculator)

def salary_calculator(emp_name, base_salary=20000, *allowances, **deductions):
    
    print("\n----- SALARY DETAILS -----")
    print("Employee Name:", emp_name)
    
    total_salary = base_salary + sum(allowances)
    
    total_deduction = sum(deductions.values())
    
    final_salary = total_salary - total_deduction
    
    print("Base Salary:", base_salary)
    print("Allowances:", allowances)
    print("Deductions:", deductions)
    print("Final Salary:", final_salary)
```

```python
#Function Call:

salary_calculator(
    "Ritesh",
    30000,
    5000, 2000,      # Allowances
    tax=3000,
    insurance=1000
)
```

```python
#-- REAL LIFE EXAMPLE -- 

#Real Industry Example: E-Commerce Order Processing System

#In real companies like Amazon, Flipkart, etc., order processing functions handle:
-- Customer info
-- Default delivery method
-- Multiple products
-- Extra order details
```

```python
#Complete Industry-Level Function


def process_order(customer_name, customer_id,
                  payment_method="Cash on Delivery",
                  *products,
                  **order_details):
    
    print("========== ORDER SUMMARY ==========")
    
    # 1️⃣ Positional Arguments
    print("Customer Name:", customer_name)
    print("Customer ID:", customer_id)
    
    # 2️⃣ Default Argument
    print("Payment Method:", payment_method)
    
    # 3️⃣ Variable-Length Arguments (*args)
    print("\nProducts Ordered:")
    total_amount = 0
    
    for product in products:
        print("-", product["name"], "| Price:", product["price"])
        total_amount += product["price"]
    
    # 4️⃣ Keyword Variable-Length Arguments (**kwargs)
    print("\nAdditional Order Details:")
    for key, value in order_details.items():
        print(key, ":", value)
    
    print("\nTotal Amount:", total_amount)
    print("===================================")
```

```python
#Function Call (Real Use Case)

process_order(
    "Ritesh Pethkar",      # Positional
    101,                   # Positional
    
    "UPI",                 # Default overridden
    
    # *args (Multiple products as dictionaries)
    {"name": "Laptop", "price": 60000},
    {"name": "Mouse", "price": 1500},
    {"name": "Keyboard", "price": 2500},
    
    # **kwargs (Extra order info)
    delivery_address="Pune, Maharashtra",
    delivery_type="Express",
    discount=500
)
```
