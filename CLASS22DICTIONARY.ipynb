```python
#-- What is a Dictionary?

-- A dictionary in Python is a built-in data type used to store data in key–value pairs.


-Mutable (can be changed)
-Unordered (in older versions before Python 3.7; now it maintains insertion order)
-Indexed by keys (not by position like lists)
-Keys must be unique
-Keys must be immutable types (int, str, tuple, etc.)
```

```python
#Syntax of Dictionary
dict_name = {
    key1: value1,
    key2: value2,
    key3: value3
}
```

```python
#Example:
student = {
    "name": "Vishal",
    "age": 21,
    "course": "Data Analytics"
}

print(student)
```

```python
#Characteristics of Dictionary
Feature	                  Description
Key-Value Pair	          Data stored as key : value
Mutable	                  You can add, update, delete
Unique Keys	              Duplicate keys not allowed
Any Data Type             Values can be any type
```

```python
#Creating Dictionary in Different Ways


#1. Using Curly Braces
d = {"a": 1, "b": 2}
```

```python
#️2. Using dict() Constructor
d = dict(a=1, b=2)
```

```python
#3. Empty Dictionary
d = {}
```

```python
#Accessing Dictionary Elements

#Using Key
student = {"name": "Vishal", "age": 21}
print(student["name"])
print(student["age"])
```

```python
#Using get() Method

print(student.get("age"))
```

```python
#Adding & Updating Values


#Add New Key
student["city"] = "Pune"


#Update Value
student["age"] = 22
```

```python
print(student)
```

```python
#Removing Elements
️

#1.pop()
student.pop("age")
print(student)

#2.del Keyword
del student["city"]
print(student)
️
#3.popitem() (removes last item)
student.popitem()
print(student)
```

```python
#-- Dictionary Methods (Important for Exams)
Method	Description
keys()	Returns all keys
values()	Returns all values
items()	Returns key-value pairs
update()	Updates dictionary
clear()	Removes all items
copy()	Returns copy
```

```python
#-- Example:
d = {"a": 10, "b": 20}

print(d.keys())
print(d.values())
print(d.items())
```

```python
#-- Looping Through Dictionary

#✔ Loop Through Keys
for key in d:
    print(key)
#✔ Loop Through Values
for value in d.values():
    print(value)
#✔ Loop Through Key-Value
for key, value in d.items():
    print(key, value)
```

```python
#-- Nested Dictionary

#Dictionary inside dictionary.

students = {
    "s1": {"name": "Vishal", "age": 21},
    "s2": {"name": "Rahul", "age": 22}
}

print(students["s1"]["name"])
print(students["s1"]["age"])
```

```python
#Dictionary vs List
Dictionary	                    List
Uses keys	                 Uses index
key:value pair	             Only values
Unordered (logically)	      Ordered
Faster search by key	    Slower search
```

```python
#1. What is Dictionary?

# 👉 Collection of key : value pairs
# 👉 Mutable (can change)
# 👉 Keys must be unique & immutable

d = {"name": "Vishal", "age": 21}
```

```python
#2. Create Dictionary
# Normal
d = {"a": 1, "b": 2}

# Empty
d = {}

# Using dict()
d = dict(a=1, b=2)
```

```python
#3. Access Values
d["a"]          # Direct access (Error if not found)
d.get("a")      # Safe access (returns None if not found)
```

```python
#4. Add / Update
d["c"] = 3          # Add
d["a"] = 100        # Update
```

```python
#5. Delete Items
del d["a"]          # Delete specific
d.pop("b")          # Remove and return value
d.popitem()         # Remove last item
d.clear()           # Remove all
```

```python
#6. Important Methods
d.keys()        # All keys
d.values()      # All values
d.items()       # Key-value pairs
d.update({"x":10})  # Update dictionary
d.copy()        # Copy dictionary
```

```python
#7. Looping
for k in d:                 # Keys
    print(k)

for v in d.values():        # Values
    print(v)

for k, v in d.items():      # Both
    print(k, v)
```

```python
#BIG REAL-LIFE EXAMPLE OF DICTIONARY

# School Management System (Real World Project Type)

# In real life, dictionaries are used in:
-School systems
-E-commerce websites
-Banking systems
-Data Analytics projects
-APIs & JSON data
Let’s build a School Student Management System using dictionary.
```

```python
#Problem Statement

We want to:
-Add students
-Update marks
-Delete students
-Find topper
-Display all students
```

```python
# School Management System

students = {}

while True:
    print("\n===== SCHOOL MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. Update Marks")
    print("3. Delete Student")
    print("4. Display All Students")
    print("5. Find Topper")
    print("6. Exit")

    choice = input("Enter choice: ")

    # 1️⃣ Add Student
    if choice == "1":
        roll = input("Enter Roll No: ")
        name = input("Enter Name: ")
        marks = int(input("Enter Marks: "))

        students[roll] = {
            "name": name,
            "marks": marks
        }
        print("Student Added Successfully!")

    # 2️⃣ Update Marks
    elif choice == "2":
        roll = input("Enter Roll No to update: ")
        if roll in students:
            new_marks = int(input("Enter New Marks: "))
            students[roll]["marks"] = new_marks
            print("Marks Updated!")
        else:
            print("Student Not Found!")

    # 3️⃣ Delete Student
    elif choice == "3":
        roll = input("Enter Roll No to delete: ")
        if roll in students:
            students.pop(roll)
            print("Student Deleted!")
        else:
            print("Student Not Found!")

    # 4️⃣ Display All
    elif choice == "4":
        print("\nStudent Records:")
        for roll, data in students.items():
            print("Roll:", roll)
            print("Name:", data["name"])
            print("Marks:", data["marks"])
            print("------------------")

    # 5️⃣ Find Topper
    elif choice == "5":
        if students:
            topper = max(students, key=lambda x: students[x]["marks"])
            print("Topper is:", students[topper]["name"])
            print("Marks:", students[topper]["marks"])
        else:
            print("No Students Available!")

    # 6️⃣ Exit
    elif choice == "6":
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice!")
```
