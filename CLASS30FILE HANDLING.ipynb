```python
#1.Definition
File Handling in Python allows programs to create, read, write, and manage files stored on a computer.
It is used to store data permanently so it can be retrieved later essential in Data Science for reading datasets and saving results.
```

```python
#2. Key Concepts
- Python uses the built-in open() function to work with files.
- Files can be opened in different modes: read, write, append, etc.
- Always close a file after use __ or use with statement (recommended).
- File types commonly used: .txt, .csv, json, .xlsx.
- In Data Analytics, file handling is used to load and save datasets.
```

```python
#3. Syntax (File Handling in Python)


Basic Syntax
-- Functions used:
1.open()
2.read()
3.write()
4.close()
```

```python
#-- Common Modes

'r' → Read (default mode)

'w' → Write (overwrites existing content)

'a' → Append (adds data to the end of file)

'r+' → Read and Write
```

```python
#-- syntax -- 


#-- Open a File
file = open('filename.txt', 'mode')



#-- #Recommended: use 'with' statement
with open('filename.txt', 'r') as file:
    content = file.read()
    print(content)


#Explanation

-- with open() → opens the file safely
-- as file → creates a file object
-- content = file.read() → reads the file data
-- print(content) → displays the content
-- When the with block ends, the file automatically closes.
```

```python
#exmple :-- create file 

file = open("students.txt", "w")

file.write("Name: Ritesh\n")
file.write("Course: Python\n")
file.write("Age: 23\n")

file.close()

print("File created and data written successfully")
```

```python
#-- w+ --write + read both of done using (w+) fun.


-- The mode w+ in file handling (Python) means:
- w+ = Write + Read (both)
- Creates a new file if it does not exist
- Overwrites (deletes old data) if the file already exists
- Opens the file for both writing and reading
```

```python
#Example:
with open("data.txt", "w+") as f:
    f.write("Hello Python\n")
    f.seek(0)   # move cursor to beginning
    print(f.read())
```

```python
#-- r+ --  Read + Write (both)

- File must already exist ❗
- Opens file for reading and writing
- Does NOT delete old data
- Cursor starts at the beginning (position 0)
```

```python
#Example:
with open("data.txt", "r+") as f:
    print(f.read())   # read existing data
    
    f.write("Hello\n")  # write data (at end if pointer moved)
```

```python
#Important Behavior
#--Overwriting Problem
with open("data.txt", "r+") as f:
    f.write("Hi")

#This will overwrite starting characters, not append!
```

```python
#orrect Way (Append using r+)
with open("data.txt", "r+") as f:
    f.seek(0, 2)   # move to end of file
    f.write("New Line\n")
```

```python
#-- a+ -- Append + Read (both)

- Creates file if it does not exist ✅
- Opens file for reading and writing
- Does NOT delete old data
- Always writes at the end of the file
- Cursor starts at the end
```

```python
#Example:
with open("data.txt", "a+") as f:
    f.write("New Line\n")   # always added at end
    
    f.seek(0)               # move to beginning to read
    print(f.read())
```

```python
#MINI Project -- 

# 1. Advanced Student Record System (CRUD)
# Features:

# Add 
# View 
# Search 
# Update 
# Delete 
```

```python
def add_student():
    name = input("Enter name: ")
    age = input("Enter age: ")
    with open("students.txt", "a") as f:
        f.write(name + "-" + age + "\n")

def view_students():
    with open("students.txt", "r") as f:
        print(f.read())

def search_student():
    name = input("Enter name to search: ")
    with open("students.txt", "r") as f:
        for line in f:
            if name in line:
                print("Found:", line)

def update_student():
    name = input("Enter name to update: ")
    new_age = input("Enter new age: ")
    lines = []

    with open("students.txt", "r") as f:
        lines = f.readlines()

    with open("students.txt", "w") as f:
        for line in lines:
            if name in line:
                f.write(name + "-" + new_age + "\n")
            else:
                f.write(line)

def delete_student():
    name = input("Enter name to delete: ")
    lines = []

    with open("students.txt", "r") as f:
        lines = f.readlines()

    with open("students.txt", "w") as f:
        for line in lines:
            if name not in line:
                f.write(line)

while True:
    print("\n1.Add 2.View 3.Search 4.Update 5.Delete 6.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
        update_student()
    elif choice == "5":
        delete_student()
    elif choice == "6":
        break
```
